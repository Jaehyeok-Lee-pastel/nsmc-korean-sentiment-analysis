```python
# NSMC 한국어 텍스트 전처리 파이프라인
# Day 4: 전처리 + TOP 3 모델 비교 (모듈화 적용 버전)

print("✅ 라이브러리 임포트 완료!")
print(f"scikit-learn 버전: {__import__('sklearn').__version__}")

# =============================================================================
# 0. 환경 설정 (모듈화된 함수 사용)
# =============================================================================

# 머신러닝 라이브러리
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.pipeline import Pipeline

from pathlib import Path
import sys
sys.path.append(r'D:\dataAnalystBook\my_workspace\movie_review') # project_root 경로
from src.utils.setup import setup_notebook_environment

"""
setup_notebook_environment() 함수가 수행하는 작업:
1. 프로젝트 루트 자동 탐지 및 sys.path 추가
2. 한글 폰트 설정 (matplotlib에서 한글 표시)
3. pandas 표시 옵션 설정 (최대 행/열 수 조정)
4. seaborn 스타일 및 색상 팔레트 설정
5. 경고 메시지 숨김 설정
6. 주요 패키지 버전 확인 및 출력
→ 결과: 모든 노트북에서 일관된 환경 구성
"""
setup_notebook_environment()

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
# =============================================================================
# 1. 데이터 로딩 (모듈화된 함수 사용)
# =============================================================================

from src.data.data_loader import load_nsmc_data

"""
load_nsmc_data() 함수가 수행하는 작업:
1. 데이터 디렉토리 자동 탐지 (data/raw/ 폴더 찾기)
2. 다중 인코딩 시도 (utf-8, cp949, euc-kr)
3. 데이터 검증 (컬럼명, 데이터 크기, 라벨값 확인)
4. 결측값 및 데이터 품질 체크
5. 상세한 데이터 요약 정보 출력
→ 결과: 안전하고 검증된 NSMC 데이터 로딩 완료
"""
train_df, test_df = load_nsmc_data()

print(train_df.head())

# 기본 정보 확인
print("\n📋 데이터 기본 정보:")
print(f"훈련 데이터: {train_df.shape}")
print(f"테스트 데이터: {test_df.shape}")
print(f"컬럼: {train_df.columns.tolist()}")

# =============================================================================
# 2. 한국어 텍스트 전처리 (모듈화된 함수 사용)
# =============================================================================

from src.data.text_preprocessor import preprocess_text_data

"""
preprocess_text_data() 함수가 수행하는 작업:
1. 텍스트 정제 (clean_korean_text 함수 내부 호출):
   - 반복 문자 정규화 (ㅋㅋㅋ → ㅋㅋ, !!! → !!)
   - 의미없는 특수문자 제거 (감정 표현은 보존)
   - 여러 공백을 하나로 통합
   - 앞뒤 공백 제거
2. 빈 텍스트 제거 및 개수 리포트
3. 텍스트 길이 계산 및 통계 출력
4. 진행 상황 표시 (tqdm 프로그레스 바)
→ 결과: 'cleaned_text', 'text_length' 컬럼이 추가된 정제된 데이터프레임
"""

print("\n" + "="*50)
print("🔧 데이터 전처리 시작")
print("="*50)

# 훈련 데이터 전처리
print("\n📊 훈련 데이터 전처리...")
train_processed = preprocess_text_data(train_df)

# 테스트 데이터 전처리  
print("\n📊 테스트 데이터 전처리...")
test_processed = preprocess_text_data(test_df)

# 전처리 결과 확인
print("\n✅ 전처리 완료!")
print(f"훈련 데이터: {len(train_processed):,}개")
print(f"테스트 데이터: {len(test_processed):,}개")

# 전처리 예시 확인
print("\n📝 전처리 예시:")
for i in range(3):
    original = train_df.iloc[i]['document']
    cleaned = train_processed.iloc[i]['cleaned_text']
    print(f"\n예시 {i+1}:")
    print(f"원본: {original}")
    print(f"정제: {cleaned}")

# =============================================================================
# 3. 데이터 분할 및 TF-IDF 벡터화
# =============================================================================

print("\n" + "="*50)
print("🔀 데이터 분할 및 벡터화")
print("="*50)

# 텍스트와 라벨 분리
X_train_text = train_processed['cleaned_text'].values
y_train = train_processed['label'].values
X_test_text = test_processed['cleaned_text'].values  
y_test = test_processed['label'].values

from sklearn.model_selection import train_test_split

"""
train_test_split() 함수가 수행하는 작업:
1. 훈련 데이터를 train/validation으로 8:2 분할
2. stratify=y_train으로 클래스 비율 유지
3. random_state=42로 재현 가능한 분할
→ 결과: 균형잡힌 train/validation 데이터셋 생성
"""

# 훈련 데이터를 train/validation으로 분할
X_train, X_val, y_train_split, y_val = train_test_split(
    X_train_text, y_train, 
    test_size=0.2, 
    random_state=42, 
    stratify=y_train
)

print(f"📊 데이터 분할 결과:")
print(f"훈련: {len(X_train):,}개")
print(f"검증: {len(X_val):,}개") 
print(f"테스트: {len(X_test_text):,}개")

# TF-IDF 벡터화
print(f"\n🔢 TF-IDF 벡터화 중...")

from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

"""
TfidfVectorizer 설정 및 변환 과정:
1. TF-IDF 설정:
   - max_features=10000: 최대 10,000개 특성으로 제한
   - min_df=2: 최소 2개 문서에 등장하는 단어만 사용
   - max_df=0.95: 95% 이상 문서에 등장하는 흔한 단어 제거
   - ngram_range=(1,2): 단어(1-gram)와 2단어 조합(2-gram) 사용
   - sublinear_tf=True: TF 값에 로그 스케일링 적용

2. 변환 과정:
   - fit_transform(X_train): 훈련 데이터로 어휘 학습 + 벡터 변환
   - transform(X_val, X_test): 학습된 어휘로 검증/테스트 데이터 변환
   
→ 결과: 희소 행렬 형태의 TF-IDF 벡터 (문서 x 단어 특성)
"""

# TF-IDF 설정 (한국어 특화)
from src.features.tfidf_vectorizer import create_tfidf_vectorizer, show_tfidf_stats

# TF-IDF 벡터라이저 생성
tfidf = create_tfidf_vectorizer()

# 훈련 데이터로 TF-IDF 학습 및 변환
X_train_tfidf = tfidf.fit_transform(X_train)
X_val_tfidf = tfidf.transform(X_val)
X_test_tfidf = tfidf.transform(X_test_text)

print(f"✅ TF-IDF 벡터화 완료!")
print(f"🔢 특성 수: {X_train_tfidf.shape[1]:,}개")
print(f"📊 희소성: {(1 - X_train_tfidf.nnz / np.prod(X_train_tfidf.shape)) * 100:.1f}%")

# 주요 특성 확인
feature_names = tfidf.get_feature_names_out()
print(f"\n📝 주요 특성 예시 (처음 20개):")
print(feature_names[:20])

print(f"\n🎯 전처리 단계 완료! 이제 모델 학습을 시작합니다.")
print(f"⏰ 소요 시간: 약 5-10분 예상")

# =============================================================================
# 4. 모듈화 효과 요약
# =============================================================================

print(f"\n📋 모듈화된 함수들이 수행한 작업 요약:")
print(f"✅ setup_notebook_environment(): 환경 설정 자동화")
print(f"   → 프로젝트 경로, 폰트, pandas 옵션, 패키지 버전 확인")
print(f"✅ load_nsmc_data(): 안전한 데이터 로딩")
print(f"   → 자동 경로 탐지, 인코딩 처리, 데이터 검증")
print(f"✅ preprocess_text_data(): 한국어 특화 전처리")
print(f"   → 텍스트 정제, 빈 값 제거, 길이 계산, 통계 출력")

print(f"\n🎯 최종 준비 완료:")
print(f"  - 정제된 텍스트 데이터: {len(train_processed):,} + {len(test_processed):,}개")
print(f"  - TF-IDF 벡터: {X_train_tfidf.shape} (훈련)")
print(f"  - 검증/테스트 벡터: {X_val_tfidf.shape}, {X_test_tfidf.shape}")
print(f"  - 다음 단계: 머신러닝 모델 학습 (나이브베이즈, 로지스틱회귀, SVM)")
```

```python
# =============================================================================
# 나이브 베이즈 (Naive Bayes) 상세 분석
# =============================================================================

import time
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

print("🧠 나이브 베이즈 (Naive Bayes) 상세 분석 시작!")
print("="*60)

# =============================================================================
# 1. 나이브 베이즈 모델 정의 및 설명
# =============================================================================

print("📚 나이브 베이즈 모델 설정:")
print("▪️ 알고리즘: MultinomialNB (다항 나이브 베이즈)")
print("▪️ 특징: 텍스트 분류에 특화된 버전")
print("▪️ 가정: 단어들이 서로 독립적")
print("▪️ 장점: 빠른 학습, 높은 성능, 해석 가능")

# 나이브 베이즈 모델 생성
nb_model = MultinomialNB(alpha=1.0)  # alpha: 라플라스 스무딩 (기본값 1.0)

print(f"\n⚙️ 하이퍼파라미터:")
print(f"▪️ alpha (스무딩): {nb_model.alpha}")
print(f"   → 0이 나오지 않는 단어에 대한 처리 (0확률 방지)")

# =============================================================================
# 2. 교차 검증으로 안정적 성능 측정
# =============================================================================

print(f"\n🔬 5-fold 교차 검증 실행...")
print(f"📊 데이터: {X_train_tfidf.shape[0]:,}개 샘플 × {X_train_tfidf.shape[1]:,}개 특성")

start_time = time.time()

# 5-fold 교차 검증 실행
cv_scores = cross_val_score(
    nb_model, 
    X_train_tfidf, 
    y_train_split, 
    cv=5, 
    scoring='accuracy'
)

cv_time = time.time() - start_time

print(f"✅ 교차 검증 완료!")
print(f"⏰ 소요 시간: {cv_time:.2f}초")
print(f"\n📊 교차 검증 결과:")
print(f"▪️ 평균 정확도: {cv_scores.mean():.4f}")
print(f"▪️ 표준편차: {cv_scores.std():.4f}")
print(f"▪️ 95% 신뢰구간: {cv_scores.mean():.4f} ± {cv_scores.std()*2:.4f}")

print(f"\n📈 각 fold별 성능:")
for i, score in enumerate(cv_scores, 1):
    print(f"   Fold {i}: {score:.4f}")

# =============================================================================
# 3. 전체 훈련 데이터로 최종 학습
# =============================================================================

print(f"\n🎓 전체 훈련 데이터로 최종 학습...")

start_time = time.time()
nb_model.fit(X_train_tfidf, y_train_split)
train_time = time.time() - start_time

print(f"✅ 학습 완료! 소요 시간: {train_time:.3f}초")

# 검증 데이터로 예측
val_predictions = nb_model.predict(X_val_tfidf)
val_probabilities = nb_model.predict_proba(X_val_tfidf)

# 성능 계산
val_accuracy = accuracy_score(y_val, val_predictions)

print(f"\n📊 검증 데이터 성능:")
print(f"▪️ 정확도: {val_accuracy:.4f}")

# 상세 성능 분석
report = classification_report(y_val, val_predictions, 
                             target_names=['부정', '긍정'], 
                             output_dict=True)

print(f"▪️ 정밀도 (긍정): {report['긍정']['precision']:.4f}")
print(f"▪️ 재현율 (긍정): {report['긍정']['recall']:.4f}")
print(f"▪️ F1-Score (긍정): {report['긍정']['f1-score']:.4f}")

print(f"▪️ 정밀도 (부정): {report['부정']['precision']:.4f}")
print(f"▪️ 재현율 (부정): {report['부정']['recall']:.4f}")
print(f"▪️ F1-Score (부정): {report['부정']['f1-score']:.4f}")

# =============================================================================
# 4. 나이브 베이즈의 핵심: 단어별 확률 분석
# =============================================================================

print(f"\n🔍 나이브 베이즈 핵심 분석: 단어별 확률")
print("="*60)

# 특성명 가져오기
feature_names = tfidf.get_feature_names_out()

# 각 클래스별 로그 확률 (부정=0, 긍정=1)
log_prob_neg = nb_model.feature_log_prob_[0]  # 부정 클래스
log_prob_pos = nb_model.feature_log_prob_[1]  # 긍정 클래스

# 로그 확률 차이 계산 (긍정 - 부정)
log_prob_diff = log_prob_pos - log_prob_neg

print(f"📈 가장 강한 긍정 신호 단어들 (TOP 15):")
top_positive_idx = np.argsort(log_prob_diff)[-15:]
for i, idx in enumerate(top_positive_idx[::-1], 1):
    word = feature_names[idx]
    diff = log_prob_diff[idx]
    pos_prob = np.exp(log_prob_pos[idx])
    neg_prob = np.exp(log_prob_neg[idx])
    print(f"   {i:2d}. {word:15} | 차이: {diff:6.3f} | 긍정확률: {pos_prob:.6f} | 부정확률: {neg_prob:.6f}")

print(f"\n📉 가장 강한 부정 신호 단어들 (TOP 15):")
top_negative_idx = np.argsort(log_prob_diff)[:15]
for i, idx in enumerate(top_negative_idx, 1):
    word = feature_names[idx]
    diff = log_prob_diff[idx]
    pos_prob = np.exp(log_prob_pos[idx])
    neg_prob = np.exp(log_prob_neg[idx])
    print(f"   {i:2d}. {word:15} | 차이: {diff:6.3f} | 긍정확률: {pos_prob:.6f} | 부정확률: {neg_prob:.6f}")

# =============================================================================
# 5. 실제 예측 과정 시연
# =============================================================================

print(f"\n🎯 실제 예측 과정 시연:")
print("="*60)

# 검증 데이터에서 몇 개 샘플 선택
sample_indices = [0, 100, 500, 1000, 2000]

for i, idx in enumerate(sample_indices):
    actual_text = X_val[idx]
    actual_label = y_val[idx]
    predicted_label = val_predictions[idx]
    predicted_proba = val_probabilities[idx]
    
    print(f"\n예시 {i+1}:")
    print(f"원본 텍스트: {actual_text}")
    print(f"실제 라벨: {'긍정' if actual_label == 1 else '부정'}")
    print(f"예측 라벨: {'긍정' if predicted_label == 1 else '부정'}")
    print(f"예측 확률: 부정 {predicted_proba[0]:.3f}, 긍정 {predicted_proba[1]:.3f}")
    
    # 예측이 맞았는지 확인
    if actual_label == predicted_label:
        print("✅ 정답!")
    else:
        print("❌ 오답!")

# =============================================================================
# 6. 혼동 행렬 시각화
# =============================================================================

print(f"\n📊 나이브 베이즈 혼동 행렬:")

# 혼동 행렬 계산
cm = confusion_matrix(y_val, val_predictions)

# 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(cm, 
            annot=True, 
            fmt='d', 
            cmap='Blues',
            xticklabels=['부정', '긍정'],
            yticklabels=['부정', '긍정'],
            cbar_kws={'label': '샘플 수'})

plt.title(f'나이브 베이즈 혼동 행렬\n정확도: {val_accuracy:.4f}', 
          fontsize=14, fontweight='bold')
plt.xlabel('예측 라벨', fontsize=12)
plt.ylabel('실제 라벨', fontsize=12)

# 각 셀에 백분율 추가
total = cm.sum()
for i in range(2):
    for j in range(2):
        percentage = cm[i, j] / total * 100
        plt.text(j + 0.5, i + 0.7, f'({percentage:.1f}%)', 
                ha='center', va='center', fontsize=10, color='gray')

plt.tight_layout()
plt.show()

# =============================================================================
# 7. 나이브 베이즈 성능 요약
# =============================================================================

print(f"\n🎯 나이브 베이즈 최종 성능 요약:")
print("="*60)

print(f"📊 핵심 성능 지표:")
print(f"▪️ 교차 검증 정확도: {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")
print(f"▪️ 검증 데이터 정확도: {val_accuracy:.4f}")
print(f"▪️ 전체 F1-Score: {report['macro avg']['f1-score']:.4f}")

print(f"\n⚡ 효율성 지표:")
print(f"▪️ 교차 검증 시간: {cv_time:.2f}초")
print(f"▪️ 최종 학습 시간: {train_time:.3f}초")
print(f"▪️ 초당 처리 샘플: {len(X_train)/train_time:,.0f}개")

print(f"\n🎯 목표 달성 여부:")
target_75 = 0.75
target_80 = 0.80

if val_accuracy >= target_80:
    print(f"🚀 우수! 80% 목표 달성: {val_accuracy:.4f} >= {target_80:.4f}")
elif val_accuracy >= target_75:
    print(f"✅ 성공! 75% 목표 달성: {val_accuracy:.4f} >= {target_75:.4f}")
else:
    print(f"⚠️  목표 미달성: {val_accuracy:.4f} < {target_75:.4f}")

print(f"\n💡 나이브 베이즈 특징 요약:")
print(f"✅ 장점: 매우 빠른 학습 ({train_time:.3f}초)")
print(f"✅ 장점: 완벽한 해석 가능성 (단어별 확률 확인)")
print(f"✅ 장점: 텍스트 분류에 특화된 알고리즘")
print(f"✅ 장점: 메모리 효율적")
print(f"⚠️  단점: 단어 독립성 가정의 한계")
print(f"⚠️  단점: 복잡한 단어 조합 패턴 놓칠 수 있음")

print(f"\n🏁 나이브 베이즈 분석 완료!")
print(f"⏰ 완료 시간: {datetime.now().strftime('%H:%M:%S')}")

# 결과를 다음 단계를 위해 저장
nb_results = {
    'model': nb_model,
    'cv_scores': cv_scores,
    'cv_mean': cv_scores.mean(),
    'cv_std': cv_scores.std(),
    'val_accuracy': val_accuracy,
    'train_time': train_time,
    'cv_time': cv_time,
    'predictions': val_predictions,
    'probabilities': val_probabilities,
    'classification_report': report
}

print(f"💾 결과 저장 완료: nb_results 변수에 저장됨")

# =============================================================================
# 8. 모델 저장
# =============================================================================

import joblib

print(f"\n💾 나이브 베이즈 모델 저장...")

# 모델 저장
joblib.dump(nb_model, '../../models/trained/naive_bayes_v1.pkl')

# TF-IDF 벡터라이저도 함께 저장 (Day 4 기본 설정)
joblib.dump(tfidf, '../../models/trained/tfidf_vectorizer_v1.pkl')

print(f"✅ 저장 완료:")
print(f"   ▪️ 나이브 베이즈: models/trained/naive_bayes_v1.pkl")
print(f"   ▪️ TF-IDF 벡터라이저: models/trained/tfidf_vectorizer_v1.pkl")


# =============================================================================
# 9. 예측 결과 저장
# =============================================================================

print(f"\n💾 나이브 베이즈 예측 결과 저장...")

# 예측 결과 DataFrame 생성
import pandas as pd
import os

# results 폴더 생성 (없으면)
os.makedirs('../../results/predictions', exist_ok=True)

predictions_df = pd.DataFrame({
    'actual_label': y_val,
    'predicted_label': val_predictions,
    'predicted_prob_negative': val_probabilities[:, 0],
    'predicted_prob_positive': val_probabilities[:, 1],
    'review_text': X_val,
    'correct_prediction': (y_val == val_predictions).astype(int),
    'confidence': np.max(val_probabilities, axis=1)
})

# 한글 인코딩으로 CSV 저장
predictions_df.to_csv(
    '../../results/predictions/naive_bayes_predictions.csv', 
    index=False, 
    encoding='utf-8-sig',  # Excel에서도 한글이 제대로 보이도록
    quoting=1  # 모든 문자열을 따옴표로 감싸기
)

print(f"✅ 예측 결과 저장 완료:")
print(f"   ▪️ 파일: results/predictions/naive_bayes_predictions.csv")
print(f"   ▪️ 총 {len(predictions_df):,}개 예측 결과")

# 성능 지표도 JSON으로 저장
os.makedirs('../../results/metrics', exist_ok=True)

nb_metrics = {
    'model_name': 'naive_bayes_v1',
    'model_type': 'MultinomialNB',
    'evaluation_date': datetime.now().isoformat(),
    'validation_performance': {
        'accuracy': float(val_accuracy),
        'precision_positive': float(report['긍정']['precision']),
        'recall_positive': float(report['긍정']['recall']),
        'f1_positive': float(report['긍정']['f1-score']),
        'precision_negative': float(report['부정']['precision']),
        'recall_negative': float(report['부정']['recall']),
        'f1_negative': float(report['부정']['f1-score']),
        'macro_avg_f1': float(report['macro avg']['f1-score'])
    },
    'cross_validation': {
        'cv_mean': float(cv_scores.mean()),
        'cv_std': float(cv_scores.std()),
        'cv_scores': [float(score) for score in cv_scores]
    },
    'training_info': {
        'train_time': float(train_time),
        'cv_time': float(cv_time),
        'alpha': float(nb_model.alpha)
    }
}

import json
with open('../../results/metrics/naive_bayes_metrics.json', 'w', encoding='utf-8') as f:
    json.dump(nb_metrics, f, indent=2, ensure_ascii=False)

print(f"   ▪️ 성능 지표: results/metrics/naive_bayes_metrics.json")
```

```python
# =============================================================================
# 로지스틱 회귀 (Logistic Regression) 상세 분석
# SVM 제외, 나이브 베이즈와 집중 비교
# =============================================================================

import time
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

print("📈 로지스틱 회귀 (Logistic Regression) 상세 분석 시작!")
print("="*60)

# =============================================================================
# 1. 로지스틱 회귀 모델 정의 및 설명
# =============================================================================

print("📚 로지스틱 회귀 모델 설정:")
print("▪️ 알고리즘: LogisticRegression")
print("▪️ 핵심 원리: 선형 조합 + 시그모이드 함수")
print("▪️ 장점: 완벽한 해석성, 확률 출력, 빠른 학습")
print("▪️ 목표: 각 단어별 최적 가중치 학습")

# 로지스틱 회귀 모델 생성
lr_model = LogisticRegression(
    C=1.0,               # 정규화 강도 (기본값)
    max_iter=1000,       # 최대 반복 횟수
    random_state=42,     # 재현성
    solver='liblinear'   # TF-IDF 희소 행렬에 적합한 해법
)

print(f"\n⚙️ 하이퍼파라미터:")
print(f"▪️ C: {lr_model.C} (정규화 강도)")
print(f"   → 1.0 = 균형잡힌 설정 (과적합 vs 성능)")
print(f"▪️ max_iter: {lr_model.max_iter}")
print(f"   → 최적화 알고리즘 최대 반복 횟수")
print(f"▪️ solver: {lr_model.solver}")
print(f"   → 희소 행렬에 최적화된 해법")

# =============================================================================
# 2. 교차 검증으로 안정적 성능 측정
# =============================================================================

print(f"\n🔬 5-fold 교차 검증 실행...")
print(f"📊 데이터: {X_train_tfidf.shape[0]:,}개 샘플 × {X_train_tfidf.shape[1]:,}개 특성")
print(f"🎯 작업: 10,000개 단어별 최적 가중치 탐색")

start_time = time.time()

# 5-fold 교차 검증 실행  
cv_scores = cross_val_score(
    lr_model, 
    X_train_tfidf, 
    y_train_split, 
    cv=5, 
    scoring='accuracy'
)

cv_time = time.time() - start_time

print(f"✅ 교차 검증 완료!")
print(f"⏰ 소요 시간: {cv_time:.2f}초")
print(f"\n📊 교차 검증 결과:")
print(f"▪️ 평균 정확도: {cv_scores.mean():.4f}")
print(f"▪️ 표준편차: {cv_scores.std():.4f}")
print(f"▪️ 95% 신뢰구간: {cv_scores.mean():.4f} ± {cv_scores.std()*2:.4f}")

print(f"\n📈 각 fold별 성능:")
for i, score in enumerate(cv_scores, 1):
    print(f"   Fold {i}: {score:.4f}")

# 나이브 베이즈와 교차 검증 성능 비교
nb_cv_mean = nb_results['cv_mean']
nb_cv_time = nb_results['cv_time']
lr_cv_mean = cv_scores.mean()

cv_improvement = lr_cv_mean - nb_cv_mean

print(f"\n🆚 교차 검증 성능 비교:")
print("="*60)
print(f"▪️ 나이브 베이즈: {nb_cv_mean:.4f} ± {nb_results['cv_std']:.4f} ({nb_cv_time:.2f}초)")
print(f"▪️ 로지스틱 회귀: {lr_cv_mean:.4f} ± {cv_scores.std():.4f} ({cv_time:.2f}초)")
print(f"▪️ 성능 차이: {cv_improvement:+.4f}p ({cv_improvement*100:+.2f}%p)")

if cv_improvement > 0.005:
    print(f"✅ 로지스틱 회귀가 더 우수!")
elif cv_improvement < -0.005:
    print(f"✅ 나이브 베이즈가 더 우수!")
else:
    print(f"🤝 두 모델의 성능이 비슷함")

# 시간 효율성 비교
time_ratio = cv_time / nb_cv_time
print(f"\n⚡ 학습 시간 비교:")
print(f"▪️ 로지스틱 회귀가 나이브 베이즈보다 {time_ratio:.1f}배 오래 걸림")

# =============================================================================
# 3. 전체 훈련 데이터로 최종 학습
# =============================================================================

print(f"\n🎓 전체 훈련 데이터로 최종 학습...")
print(f"🔍 작업: {X_train_tfidf.shape[0]:,}개 샘플로 10,000개 가중치 학습")

start_time = time.time()
lr_model.fit(X_train_tfidf, y_train_split)
train_time = time.time() - start_time

print(f"✅ 학습 완료! 소요 시간: {train_time:.3f}초")

# 수렴 여부 확인
if hasattr(lr_model, 'n_iter_'):
    n_iterations = lr_model.n_iter_[0]
    print(f"🔄 수렴 반복 횟수: {n_iterations}번 (최대 {lr_model.max_iter}번)")
    
    if n_iterations < lr_model.max_iter:
        print(f"✅ 정상 수렴 완료 (여유분: {lr_model.max_iter - n_iterations}번)")
    else:
        print(f"⚠️  최대 반복 횟수 도달 (수렴 불확실)")

# 검증 데이터로 예측
val_predictions = lr_model.predict(X_val_tfidf)
val_probabilities = lr_model.predict_proba(X_val_tfidf)

# 성능 계산
val_accuracy = accuracy_score(y_val, val_predictions)

print(f"\n📊 검증 데이터 성능:")
print(f"▪️ 정확도: {val_accuracy:.4f}")

# 상세 성능 분석
report = classification_report(y_val, val_predictions, 
                             target_names=['부정', '긍정'], 
                             output_dict=True)

print(f"▪️ 정밀도 (긍정): {report['긍정']['precision']:.4f}")
print(f"▪️ 재현율 (긍정): {report['긍정']['recall']:.4f}")
print(f"▪️ F1-Score (긍정): {report['긍정']['f1-score']:.4f}")

print(f"▪️ 정밀도 (부정): {report['부정']['precision']:.4f}")
print(f"▪️ 재현율 (부정): {report['부정']['recall']:.4f}")
print(f"▪️ F1-Score (부정): {report['부정']['f1-score']:.4f}")

# 나이브 베이즈와 최종 성능 비교
nb_val_accuracy = nb_results['val_accuracy']
nb_train_time = nb_results['train_time']
final_improvement = val_accuracy - nb_val_accuracy

print(f"\n🆚 최종 검증 성능 비교:")
print("="*60)
print(f"▪️ 나이브 베이즈: {nb_val_accuracy:.4f} (학습: {nb_train_time:.3f}초)")
print(f"▪️ 로지스틱 회귀: {val_accuracy:.4f} (학습: {train_time:.3f}초)")
print(f"▪️ 성능 차이: {final_improvement:+.4f}p ({final_improvement*100:+.2f}%p)")
print(f"▪️ 시간 차이: {train_time/nb_train_time:.1f}배 더 오래")

# =============================================================================
# 4. 로지스틱 회귀의 핵심: 가중치 분석
# =============================================================================

print(f"\n🔍 로지스틱 회귀 핵심 분석: 단어별 가중치")
print("="*60)

# 특성명과 가중치 추출
feature_names = tfidf.get_feature_names_out()
lr_coef = lr_model.coef_[0]

print(f"📊 가중치 통계:")
print(f"▪️ 총 특성 수: {len(lr_coef):,}개")
print(f"▪️ 평균 가중치: {lr_coef.mean():.6f}")
print(f"▪️ 표준편차: {lr_coef.std():.6f}")
print(f"▪️ 가중치 범위: {lr_coef.min():.3f} ~ {lr_coef.max():.3f}")
print(f"▪️ 0에 가까운 가중치: {np.sum(np.abs(lr_coef) < 0.01):,}개 ({np.sum(np.abs(lr_coef) < 0.01)/len(lr_coef)*100:.1f}%)")

# 가장 높은 가중치 (긍정 기여)
top_positive_idx = np.argsort(lr_coef)[-20:]
print(f"\n📈 긍정에 가장 기여하는 단어들 (TOP 20):")
for i, idx in enumerate(top_positive_idx[::-1], 1):
    word = feature_names[idx]
    weight = lr_coef[idx]
    print(f"   {i:2d}. {word:25} | 가중치: {weight:7.4f}")

# 가장 낮은 가중치 (부정 기여)
top_negative_idx = np.argsort(lr_coef)[:20]
print(f"\n📉 부정에 가장 기여하는 단어들 (TOP 20):")
for i, idx in enumerate(top_negative_idx, 1):
    word = feature_names[idx]
    weight = lr_coef[idx]
    print(f"   {i:2d}. {word:25} | 가중치: {weight:7.4f}")

# =============================================================================
# 5. 나이브 베이즈와 로지스틱 회귀 특성 중요도 비교
# =============================================================================

print(f"\n🆚 나이브 베이즈 vs 로지스틱 회귀 특성 중요도 비교:")
print("="*60)

# 나이브 베이즈 결과 가져오기
nb_model = nb_results['model']
nb_log_prob_diff = nb_model.feature_log_prob_[1] - nb_model.feature_log_prob_[0]

# 각 모델의 상위 15개 긍정/부정 단어 추출
top_nb_positive = set(feature_names[np.argsort(nb_log_prob_diff)[-15:]])
top_lr_positive = set(feature_names[np.argsort(lr_coef)[-15:]])
top_nb_negative = set(feature_names[np.argsort(nb_log_prob_diff)[:15]])
top_lr_negative = set(feature_names[np.argsort(lr_coef)[:15]])

# 공통 중요 단어 찾기
common_positive = top_nb_positive.intersection(top_lr_positive)
common_negative = top_nb_negative.intersection(top_lr_negative)

print(f"📈 두 모델이 공통으로 중요하게 본 긍정 단어들:")
if common_positive:
    for word in sorted(common_positive):
        idx = list(feature_names).index(word)
        nb_score = nb_log_prob_diff[idx]
        lr_weight = lr_coef[idx]
        print(f"   {word:25} | NB점수: {nb_score:6.3f} | LR가중치: {lr_weight:6.3f}")
else:
    print("   공통 단어 없음")

print(f"\n📉 두 모델이 공통으로 중요하게 본 부정 단어들:")
if common_negative:
    for word in sorted(common_negative):
        idx = list(feature_names).index(word)
        nb_score = nb_log_prob_diff[idx]
        lr_weight = lr_coef[idx]
        print(f"   {word:25} | NB점수: {nb_score:6.3f} | LR가중치: {lr_weight:6.3f}")
else:
    print("   공통 단어 없음")

# 각 모델만의 고유 중요 단어
nb_only_positive = top_nb_positive - top_lr_positive
lr_only_positive = top_lr_positive - top_nb_positive

if nb_only_positive:
    print(f"\n🔵 나이브 베이즈만 중요하게 본 긍정 단어들:")
    for word in sorted(nb_only_positive):
        idx = list(feature_names).index(word)
        print(f"   {word:25} | NB점수: {nb_log_prob_diff[idx]:6.3f}")

if lr_only_positive:
    print(f"\n🔴 로지스틱 회귀만 중요하게 본 긍정 단어들:")
    for word in sorted(lr_only_positive):
        idx = list(feature_names).index(word)
        print(f"   {word:25} | LR가중치: {lr_coef[idx]:6.3f}")

# =============================================================================
# 6. 실제 예측 과정 시연 (2개 모델 비교)
# =============================================================================

print(f"\n🎯 실제 예측 과정 시연 (나이브 베이즈 vs 로지스틱 회귀):")
print("="*60)

# 검증 데이터에서 샘플 선택
sample_indices = [0, 100, 500, 1000, 2000, 5000, 10000]

for i, idx in enumerate(sample_indices):
    actual_text = X_val[idx]
    actual_label = y_val[idx]
    
    # 각 모델의 예측
    nb_pred = nb_results['predictions'][idx]
    nb_proba = nb_results['probabilities'][idx]
    
    lr_pred = val_predictions[idx]
    lr_proba = val_probabilities[idx]
    
    print(f"\n예시 {i+1}:")
    print(f"텍스트: {actual_text}")
    print(f"실제: {'긍정' if actual_label == 1 else '부정'}")
    
    print(f"NB 예측: {'긍정' if nb_pred == 1 else '부정'} | 확률: 부정 {nb_proba[0]:.3f}, 긍정 {nb_proba[1]:.3f}")
    print(f"LR 예측: {'긍정' if lr_pred == 1 else '부정'} | 확률: 부정 {lr_proba[0]:.3f}, 긍정 {lr_proba[1]:.3f}")
    
    # 정답 여부 확인
    nb_correct = "✅" if actual_label == nb_pred else "❌"
    lr_correct = "✅" if actual_label == lr_pred else "❌"
    
    print(f"결과: NB {nb_correct} | LR {lr_correct}")
    
    # 두 모델의 예측이 다른 경우
    if nb_pred != lr_pred:
        print("🤔 두 모델의 예측이 다름!")
        
        # 확률 차이 분석
        nb_confidence = max(nb_proba)
        lr_confidence = max(lr_proba)
        print(f"   NB 신뢰도: {nb_confidence:.3f} | LR 신뢰도: {lr_confidence:.3f}")

# =============================================================================
# 7. 혼동 행렬 시각화 및 비교
# =============================================================================

print(f"\n📊 나이브 베이즈 vs 로지스틱 회귀 혼동 행렬 비교:")

# 혼동 행렬 계산
nb_cm = confusion_matrix(y_val, nb_results['predictions'])
lr_cm = confusion_matrix(y_val, val_predictions)

# 시각화
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle('나이브 베이즈 vs 로지스틱 회귀 성능 비교', fontsize=16, fontweight='bold')

# 나이브 베이즈 혼동 행렬
sns.heatmap(nb_cm, 
            annot=True, 
            fmt='d', 
            cmap='Blues',
            xticklabels=['부정', '긍정'],
            yticklabels=['부정', '긍정'],
            ax=axes[0])
axes[0].set_title(f'나이브 베이즈\n정확도: {nb_val_accuracy:.4f}\n학습시간: {nb_train_time:.3f}초')
axes[0].set_xlabel('예측')
axes[0].set_ylabel('실제')

# 로지스틱 회귀 혼동 행렬
sns.heatmap(lr_cm, 
            annot=True, 
            fmt='d', 
            cmap='Reds',
            xticklabels=['부정', '긍정'],
            yticklabels=['부정', '긍정'],
            ax=axes[1])
axes[1].set_title(f'로지스틱 회귀\n정확도: {val_accuracy:.4f}\n학습시간: {train_time:.3f}초')
axes[1].set_xlabel('예측')
axes[1].set_ylabel('실제')

# 각 셀에 백분율 추가
for ax, cm in zip(axes, [nb_cm, lr_cm]):
    total = cm.sum()
    for i in range(2):
        for j in range(2):
            percentage = cm[i, j] / total * 100
            ax.text(j + 0.5, i + 0.7, f'({percentage:.1f}%)', 
                   ha='center', va='center', fontsize=9, color='white')

plt.tight_layout()
plt.show()


# =============================================================================
# 8. 목표 달성 여부 및 최종 결론
# =============================================================================

print(f"\n🎯 목표 달성 여부:")
print("="*60)

target_75 = 0.75
target_80 = 0.80

print(f"📊 로지스틱 회귀 최종 성능: {val_accuracy:.4f}")

if val_accuracy >= target_80:
    print(f"🚀 우수! 80% 목표 달성: {val_accuracy:.4f} >= {target_80:.4f}")
elif val_accuracy >= target_75:
    print(f"✅ 성공! 75% 목표 달성: {val_accuracy:.4f} >= {target_75:.4f}")
    
    # 80%까지 얼마나 부족한지
    gap_to_80 = target_80 - val_accuracy
    print(f"📈 80% 목표까지: {gap_to_80:.4f}p 부족")
else:
    print(f"⚠️  목표 미달성: {val_accuracy:.4f} < {target_75:.4f}")

# 최종 모델 선택 추천
print(f"\n🏆 최종 모델 선택 추천:")
print("="*60)

models_comparison = [
    ("나이브 베이즈", nb_val_accuracy, nb_train_time, "높은 성능, 초고속 학습"),
    ("로지스틱 회귀", val_accuracy, train_time, "완벽한 해석성, 빠른 학습")
]

# 성능 기준 정렬
models_sorted = sorted(models_comparison, key=lambda x: x[1], reverse=True)

for i, (name, accuracy, time_taken, feature) in enumerate(models_sorted, 1):
    if i == 1:
        print(f"🥇 추천: {name}")
        print(f"   📊 정확도: {accuracy:.4f}")
        print(f"   ⏰ 학습시간: {time_taken:.3f}초")
        print(f"   💡 특징: {feature}")
    else:
        print(f"🥈 대안: {name}")
        print(f"   📊 정확도: {accuracy:.4f}")
        print(f"   ⏰ 학습시간: {time_taken:.3f}초")  
        print(f"   💡 특징: {feature}")

# 실무 관점에서 추천
performance_diff = abs(nb_val_accuracy - val_accuracy)
if performance_diff < 0.01:  # 성능 차이가 1% 미만
    print(f"\n💼 실무 관점 추천:")
    print(f"✅ 성능 차이가 미미하므로 ({performance_diff:.4f}p) 용도에 따라 선택")
    print(f"▪️ 실시간 처리 중요: 나이브 베이즈")
    print(f"▪️ 해석성 중요: 로지스틱 회귀")
    print(f"▪️ 균형점: 로지스틱 회귀 (해석성 + 적당한 속도)")

print(f"\n🏁 로지스틱 회귀 분석 완료!")
print(f"⏰ 완료 시간: {datetime.now().strftime('%H:%M:%S')}")

# 결과 저장
lr_results = {
    'model': lr_model,
    'cv_scores': cv_scores,
    'cv_mean': cv_scores.mean(),
    'cv_std': cv_scores.std(),
    'val_accuracy': val_accuracy,
    'train_time': train_time,
    'cv_time': cv_time,
    'predictions': val_predictions,
    'probabilities': val_probabilities,
    'classification_report': report,
    'coefficients': lr_coef,
    'n_iterations': n_iterations if 'n_iterations' in locals() else None
}

print(f"💾 결과 저장 완료: lr_results 변수에 저장됨")

# SVM 관련 참고 정보 (문서화용)
print(f"\n📋 참고: SVM 분석 결과 (문서화용)")
print("="*60)
print("🔬 SVM (Support Vector Machine):")
print("▪️ 교차 검증 정확도: 78.77% ± 0.27%")
print("▪️ 학습 시간: 2,558초 (42분)")
print("▪️ 서포트 벡터 비율: 52.7%")
print("❌ 제외 사유: 학습 시간 과다 (나이브 베이즈 대비 21,672배)")
print("❌ 성능: 나이브 베이즈보다 0.58%p 낮음")
print("❌ 효율성: 성능 향상 없이 과도한 계산 비용")
print("💡 결론: NSMC 데이터에서는 복잡한 SVM보다 단순한 모델이 더 효과적")


# =============================================================================
# 9. 로지스틱 회귀 모델 저장
# =============================================================================

import joblib

print(f"\n💾 로지스틱 회귀 모델 저장...")

# 로지스틱 회귀 모델 저장
joblib.dump(lr_model, '../../models/trained/logistic_regression_v1.pkl')

print(f"✅ 저장 완료:")
print(f"   ▪️ 로지스틱 회귀: models/trained/logistic_regression_v1.pkl")

# 성능 결과도 JSON으로 저장
import json
import os

# results 폴더 생성 (없으면)
os.makedirs('../../results/metrics', exist_ok=True)

# 로지스틱 회귀 단독 성능 지표
lr_performance = {
    'logistic_regression': {
        'cv_accuracy_mean': float(lr_results['cv_mean']),
        'cv_accuracy_std': float(lr_results['cv_std']),
        'val_accuracy': float(lr_results['val_accuracy']),
        'train_time': float(lr_results['train_time']),
        'cv_time': float(lr_results['cv_time']),
        'n_iterations': int(lr_results['n_iterations']) if lr_results['n_iterations'] else None,
        'coefficients_stats': {
            'mean': float(lr_results['coefficients'].mean()),
            'std': float(lr_results['coefficients'].std()),
            'min': float(lr_results['coefficients'].min()),
            'max': float(lr_results['coefficients'].max())
        }
    }
}

# JSON 저장
with open('../../results/metrics/logistic_regression_metrics.json', 'w', encoding='utf-8') as f:
    json.dump(lr_performance, f, indent=2, ensure_ascii=False)

print(f"   ▪️ 성능 지표: results/metrics/logistic_regression_metrics.json")

# 예측 결과도 CSV로 저장 (오류 분석용)
import pandas as pd

predictions_df = pd.DataFrame({
    'actual_label': y_val,
    'predicted_label': lr_results['predictions'],
    'predicted_prob_negative': lr_results['probabilities'][:, 0],
    'predicted_prob_positive': lr_results['probabilities'][:, 1],
    'review_text': X_val
})

predictions_df.to_csv('../../results/predictions/logistic_regression_predictions.csv', 
                    index=False, 
                    encoding='utf-8-sig',  # Excel에서도 한글이 제대로 보이도록
                    quoting=1  # 모든 문자열을 따옴표로 감싸기
)

print(f"   ▪️ 예측 결과: results/predictions/logistic_regression_predictions.csv")

print(f"\n🎯 로지스틱 회귀 모델 및 결과 저장 완료!")
```

```python
# =============================================================================
# 1단계: Soft Voting 앙상블 모델 구현
# =============================================================================

import time
from datetime import datetime
import numpy as np
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

print("🗳️ 1단계: Soft Voting 앙상블 모델 구현")
print("="*60)

# =============================================================================
# 1. 앙상블 모델 정의 및 설명
# =============================================================================

print("📚 Soft Voting 앙상블 설명:")
print("▪️ 목적: 나이브 베이즈 + 로지스틱 회귀 조합")
print("▪️ 방법: 두 모델의 예측 확률을 평균하여 결합")
print("▪️ 장점: 확신도를 고려한 더 정교한 판단")
print("▪️ 목표: 79.25% → 80%+ 정확도 향상")

# 기존 모델들 성능 확인
print(f"\n📊 기존 개별 모델 성능:")
print(f"▪️ 나이브 베이즈: {nb_results['val_accuracy']:.4f}")
print(f"▪️ 로지스틱 회귀: {lr_results['val_accuracy']:.4f}")
print(f"▪️ 성능 차이: {abs(nb_results['val_accuracy'] - lr_results['val_accuracy']):.4f}p")

# =============================================================================
# 2. Soft Voting 앙상블 생성
# =============================================================================

print(f"\n🔧 Soft Voting 앙상블 생성...")

# 앙상블 모델 정의
soft_voting_clf = VotingClassifier(
    estimators=[
        ('naive_bayes', MultinomialNB(alpha=1.0)),
        ('logistic_regression', LogisticRegression(
            C=1.0, 
            max_iter=1000, 
            solver='liblinear', 
            random_state=42
        ))
    ],
    voting='soft',  # 소프트 보팅 사용
    weights=None    # 동등한 가중치 (1:1)
)

print("✅ Soft Voting 앙상블 모델 생성 완료")
print("▪️ 구성: 나이브 베이즈 + 로지스틱 회귀")
print("▪️ 보팅: Soft (확률 평균)")
print("▪️ 가중치: 동등 (1:1)")

# =============================================================================
# 3. 교차 검증으로 성능 측정
# =============================================================================

print(f"\n🔬 Soft Voting 교차 검증 실행...")
print(f"📊 데이터: {X_train_tfidf.shape[0]:,}개 샘플")

start_time = time.time()

# 5-fold 교차 검증
cv_scores_ensemble = cross_val_score(
    soft_voting_clf, 
    X_train_tfidf, 
    y_train_split, 
    cv=5, 
    scoring='accuracy'
)

cv_time_ensemble = time.time() - start_time

print(f"✅ 교차 검증 완료! 소요 시간: {cv_time_ensemble:.2f}초")

print(f"\n📊 Soft Voting 교차 검증 결과:")
print(f"▪️ 평균 정확도: {cv_scores_ensemble.mean():.4f}")
print(f"▪️ 표준편차: {cv_scores_ensemble.std():.4f}")
print(f"▪️ 95% 신뢰구간: {cv_scores_ensemble.mean():.4f} ± {cv_scores_ensemble.std()*2:.4f}")

print(f"\n📈 각 fold별 성능:")
for i, score in enumerate(cv_scores_ensemble, 1):
    print(f"   Fold {i}: {score:.4f}")

# =============================================================================
# 4. 개별 모델 vs 앙상블 성능 비교
# =============================================================================

print(f"\n🆚 교차 검증 성능 비교:")
print("="*60)

# 성능 개선 계산
nb_cv_mean = nb_results['cv_mean']
lr_cv_mean = lr_results['cv_mean']
ensemble_cv_mean = cv_scores_ensemble.mean()

nb_improvement = ensemble_cv_mean - nb_cv_mean
lr_improvement = ensemble_cv_mean - lr_cv_mean

print(f"📊 성능 순위:")
models_cv_comparison = [
    ("Soft Voting 앙상블", ensemble_cv_mean, cv_time_ensemble),
    ("나이브 베이즈", nb_cv_mean, nb_results['cv_time']),
    ("로지스틱 회귀", lr_cv_mean, lr_results['cv_time'])
]

# 성능순 정렬
models_cv_sorted = sorted(models_cv_comparison, key=lambda x: x[1], reverse=True)

for i, (name, score, time_taken) in enumerate(models_cv_sorted, 1):
    if i == 1:
        print(f"🥇 {i}위. {name}: {score:.4f} ({time_taken:.2f}초)")
    elif i == 2:
        print(f"🥈 {i}위. {name}: {score:.4f} ({time_taken:.2f}초)")
    else:
        print(f"🥉 {i}위. {name}: {score:.4f} ({time_taken:.2f}초)")

print(f"\n📈 앙상블 개선 효과:")
print(f"▪️ vs 나이브 베이즈: {nb_improvement:+.4f}p ({nb_improvement*100:+.2f}%p)")
print(f"▪️ vs 로지스틱 회귀: {lr_improvement:+.4f}p ({lr_improvement*100:+.2f}%p)")

if ensemble_cv_mean > max(nb_cv_mean, lr_cv_mean):
    print(f"✅ 앙상블이 개별 모델들보다 우수한 성능!")
elif abs(ensemble_cv_mean - max(nb_cv_mean, lr_cv_mean)) < 0.001:
    print(f"🤝 앙상블이 최고 개별 모델과 동등한 성능")
else:
    print(f"⚠️  앙상블이 개별 모델보다 낮은 성능 (과적합 가능성)")

# =============================================================================
# 5. 최종 학습 및 검증 데이터 평가
# =============================================================================

print(f"\n🎓 전체 훈련 데이터로 최종 학습...")

start_time = time.time()
soft_voting_clf.fit(X_train_tfidf, y_train_split)
train_time_ensemble = time.time() - start_time

print(f"✅ 학습 완료! 소요 시간: {train_time_ensemble:.3f}초")

# 검증 데이터로 예측
ensemble_predictions = soft_voting_clf.predict(X_val_tfidf)
ensemble_probabilities = soft_voting_clf.predict_proba(X_val_tfidf)

# 성능 계산
ensemble_accuracy = accuracy_score(y_val, ensemble_predictions)

print(f"\n📊 검증 데이터 성능:")
print(f"▪️ 정확도: {ensemble_accuracy:.4f}")

# 상세 성능 분석
ensemble_report = classification_report(y_val, ensemble_predictions, 
                                      target_names=['부정', '긍정'], 
                                      output_dict=True)

print(f"▪️ 정밀도 (긍정): {ensemble_report['긍정']['precision']:.4f}")
print(f"▪️ 재현율 (긍정): {ensemble_report['긍정']['recall']:.4f}")
print(f"▪️ F1-Score (긍정): {ensemble_report['긍정']['f1-score']:.4f}")

print(f"▪️ 정밀도 (부정): {ensemble_report['부정']['precision']:.4f}")
print(f"▪️ 재현율 (부정): {ensemble_report['부정']['recall']:.4f}")
print(f"▪️ F1-Score (부정): {ensemble_report['부정']['f1-score']:.4f}")

# =============================================================================
# 6. 최종 성능 비교 (검증 데이터)
# =============================================================================

print(f"\n🆚 최종 검증 데이터 성능 비교:")
print("="*60)

# 개별 모델 성능
nb_val_accuracy = nb_results['val_accuracy']
lr_val_accuracy = lr_results['val_accuracy']

# 개선 효과 계산
final_nb_improvement = ensemble_accuracy - nb_val_accuracy
final_lr_improvement = ensemble_accuracy - lr_val_accuracy

models_final_comparison = [
    ("Soft Voting 앙상블", ensemble_accuracy, train_time_ensemble),
    ("나이브 베이즈", nb_val_accuracy, nb_results['train_time']),
    ("로지스틱 회귀", lr_val_accuracy, lr_results['train_time'])
]

# 성능순 정렬
models_final_sorted = sorted(models_final_comparison, key=lambda x: x[1], reverse=True)

print(f"📊 최종 성능 순위:")
for i, (name, accuracy, time_taken) in enumerate(models_final_sorted, 1):
    if i == 1:
        print(f"🥇 {i}위. {name}: {accuracy:.4f} (학습: {time_taken:.3f}초)")
    elif i == 2:
        print(f"🥈 {i}위. {name}: {accuracy:.4f} (학습: {time_taken:.3f}초)")
    else:
        print(f"🥉 {i}위. {name}: {accuracy:.4f} (학습: {time_taken:.3f}초)")

print(f"\n📈 최종 앙상블 개선 효과:")
print(f"▪️ vs 나이브 베이즈: {final_nb_improvement:+.4f}p ({final_nb_improvement*100:+.2f}%p)")
print(f"▪️ vs 로지스틱 회귀: {final_lr_improvement:+.4f}p ({final_lr_improvement*100:+.2f}%p)")

# =============================================================================
# 7. 개별 모델 예측과 앙상블 예측 비교
# =============================================================================

print(f"\n🔍 개별 모델 vs 앙상블 예측 비교 (샘플 분석):")
print("="*60)

# 검증 데이터에서 몇 개 샘플 선택
sample_indices = [0, 100, 500, 1000, 2000]

for i, idx in enumerate(sample_indices):
    actual_text = X_val[idx]
    actual_label = y_val[idx]
    
    # 개별 모델 예측
    nb_pred = nb_results['predictions'][idx]
    nb_proba = nb_results['probabilities'][idx]
    
    lr_pred = lr_results['predictions'][idx]
    lr_proba = lr_results['probabilities'][idx]
    
    # 앙상블 예측
    ensemble_pred = ensemble_predictions[idx]
    ensemble_proba = ensemble_probabilities[idx]
    
    print(f"\n예시 {i+1}:")
    print(f"텍스트: {actual_text}")
    print(f"실제: {'긍정' if actual_label == 1 else '부정'}")
    
    print(f"NB:       {'긍정' if nb_pred == 1 else '부정'} | 확률: {nb_proba[1]:.3f}")
    print(f"LR:       {'긍정' if lr_pred == 1 else '부정'} | 확률: {lr_proba[1]:.3f}")
    print(f"앙상블:   {'긍정' if ensemble_pred == 1 else '부정'} | 확률: {ensemble_proba[1]:.3f}")
    
    # 평균 계산 확인
    avg_proba = (nb_proba[1] + lr_proba[1]) / 2
    print(f"확률 평균: {avg_proba:.3f} (앙상블과 일치: {abs(avg_proba - ensemble_proba[1]) < 0.001})")
    
    # 정답 여부 확인
    nb_correct = "✅" if actual_label == nb_pred else "❌"
    lr_correct = "✅" if actual_label == lr_pred else "❌"
    ensemble_correct = "✅" if actual_label == ensemble_pred else "❌"
    
    print(f"결과: NB {nb_correct} | LR {lr_correct} | 앙상블 {ensemble_correct}")
    
    # 앙상블 효과 분석
    individual_corrects = sum([actual_label == nb_pred, actual_label == lr_pred])
    if actual_label == ensemble_pred and individual_corrects < 2:
        print("🎯 앙상블 효과: 개별 모델 실수를 보완!")
    elif actual_label != ensemble_pred and individual_corrects == 2:
        print("⚠️  앙상블 부작용: 개별 모델 정답을 오답으로")

# =============================================================================
# 8. 목표 달성 여부 확인
# =============================================================================

print(f"\n🎯 목표 달성 여부:")
print("="*60)

target_80 = 0.80
target_81 = 0.81

print(f"📊 Soft Voting 앙상블 최종 성능: {ensemble_accuracy:.4f}")

if ensemble_accuracy >= target_81:
    print(f"🚀 우수! 81% 목표 달성: {ensemble_accuracy:.4f} >= {target_81:.4f}")
elif ensemble_accuracy >= target_80:
    print(f"✅ 성공! 80% 목표 달성: {ensemble_accuracy:.4f} >= {target_80:.4f}")
    gap_to_81 = target_81 - ensemble_accuracy
    print(f"📈 81% 목표까지: {gap_to_81:.4f}p 부족")
else:
    print(f"⚠️  80% 목표 미달성: {ensemble_accuracy:.4f} < {target_80:.4f}")
    gap_to_80 = target_80 - ensemble_accuracy
    print(f"📈 80% 목표까지: {gap_to_80:.4f}p 부족")

print(f"\n🏁 1단계 Soft Voting 앙상블 분석 완료!")
print(f"⏰ 완료 시간: {datetime.now().strftime('%H:%M:%S')}")

# 결과 저장
ensemble_results = {
    'model': soft_voting_clf,
    'cv_scores': cv_scores_ensemble,
    'cv_mean': cv_scores_ensemble.mean(),
    'cv_std': cv_scores_ensemble.std(),
    'val_accuracy': ensemble_accuracy,
    'train_time': train_time_ensemble,
    'cv_time': cv_time_ensemble,
    'predictions': ensemble_predictions,
    'probabilities': ensemble_probabilities,
    'classification_report': ensemble_report
}

print(f"💾 결과 저장 완료: ensemble_results 변수에 저장됨")

# =============================================================================
# 9. 앙상블 모델 저장
# =============================================================================

import joblib

print(f"\n💾 앙상블 모델 저장...")

# 앙상블 모델 저장
joblib.dump(soft_voting_clf, '../../models/trained/ensemble_v1.pkl')

# 로지스틱 회귀 모델도 개별 저장 (비교용)
lr_model = soft_voting_clf.named_estimators_['logistic_regression']
joblib.dump(lr_model, '../../models/trained/logistic_regression_v1.pkl')

print(f"✅ 저장 완료:")
print(f"   ▪️ 앙상블 모델: models/trained/ensemble_v1.pkl")
print(f"   ▪️ 로지스틱 회귀: models/trained/logistic_regression_v1.pkl")


# =============================================================================
# 10. 앙상블 예측 결과 저장
# =============================================================================

print(f"\n💾 앙상블 예측 결과 저장...")

import pandas as pd
import os

# results 폴더 생성 (없으면)
os.makedirs('../../results/predictions', exist_ok=True)

# 앙상블 예측 결과 DataFrame 생성
ensemble_predictions_df = pd.DataFrame({
    'actual_label': y_val,
    'predicted_label': ensemble_predictions,
    'predicted_prob_negative': ensemble_probabilities[:, 0],
    'predicted_prob_positive': ensemble_probabilities[:, 1],
    'review_text': X_val,
    'correct_prediction': (y_val == ensemble_predictions).astype(int),
    'confidence': np.max(ensemble_probabilities, axis=1),
    'nb_prediction': nb_results['predictions'],
    'lr_prediction': lr_results['predictions'],
    'nb_prob_positive': nb_results['probabilities'][:, 1],
    'lr_prob_positive': lr_results['probabilities'][:, 1]
})

# 한글 인코딩으로 CSV 저장
ensemble_predictions_df.to_csv(
    '../../results/predictions/ensemble_predictions.csv', 
    index=False, 
    encoding='utf-8-sig',  # Excel에서도 한글이 제대로 보이도록
    quoting=1  # 모든 문자열을 따옴표로 감싸기
)

print(f"✅ 앙상블 예측 결과 저장 완료:")
print(f"   ▪️ 파일: results/predictions/ensemble_predictions.csv")
print(f"   ▪️ 총 {len(ensemble_predictions_df):,}개 예측 결과")
print(f"   ▪️ 개별 모델 예측도 포함 (NB, LR)")

# 앙상블 상세 성능 지표 저장
ensemble_detailed_metrics = {
    'model_name': 'ensemble_v1',
    'model_type': 'soft_voting_classifier',
    'components': ['naive_bayes', 'logistic_regression'],
    'evaluation_date': datetime.now().isoformat(),
    'validation_performance': {
        'accuracy': float(ensemble_accuracy),
        'precision_positive': float(ensemble_report['긍정']['precision']),
        'recall_positive': float(ensemble_report['긍정']['recall']),
        'f1_positive': float(ensemble_report['긍정']['f1-score']),
        'precision_negative': float(ensemble_report['부정']['precision']),
        'recall_negative': float(ensemble_report['부정']['recall']),
        'f1_negative': float(ensemble_report['부정']['f1-score']),
        'macro_avg_f1': float(ensemble_report['macro avg']['f1-score'])
    },
    'cross_validation': {
        'cv_mean': float(cv_scores_ensemble.mean()),
        'cv_std': float(cv_scores_ensemble.std()),
        'cv_scores': [float(score) for score in cv_scores_ensemble]
    },
    'training_info': {
        'train_time': float(train_time_ensemble),
        'cv_time': float(cv_time_ensemble),
        'voting_type': 'soft',
        'weights': None
    },
    'component_performance': {
        'naive_bayes_accuracy': float(nb_results['val_accuracy']),
        'logistic_regression_accuracy': float(lr_results['val_accuracy']),
        'improvement_over_nb': float(final_nb_improvement),
        'improvement_over_lr': float(final_lr_improvement)
    }
}

with open('../../results/metrics/ensemble_detailed_metrics.json', 'w', encoding='utf-8') as f:
    json.dump(ensemble_detailed_metrics, f, indent=2, ensure_ascii=False)

print(f"   ▪️ 상세 성능 지표: results/metrics/ensemble_detailed_metrics.json")



# =============================================================================
# 성능 결과 JSON 저장
# =============================================================================

import json

# 성능 지표 저장
performance_metrics = {
    'naive_bayes': {
        'cv_accuracy_mean': float(nb_results['cv_mean']),
        'cv_accuracy_std': float(nb_results['cv_std']),
        'val_accuracy': float(nb_results['val_accuracy']),
        'train_time': float(nb_results['train_time'])
    },
    'ensemble': {
        'cv_accuracy_mean': float(ensemble_results['cv_mean']),
        'cv_accuracy_std': float(ensemble_results['cv_std']),
        'val_accuracy': float(ensemble_results['val_accuracy']),
        'train_time': float(ensemble_results['train_time'])
    }
}

with open('../../results/metrics/baseline_metrics.json', 'w', encoding='utf-8') as f:
    json.dump(performance_metrics, f, indent=2, ensure_ascii=False)

print(f"✅ 성능 지표 저장: results/metrics/baseline_metrics.json")
```
