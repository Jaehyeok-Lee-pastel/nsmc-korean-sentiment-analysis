```python
# NSMC í•œêµ­ì–´ í…ìŠ¤íŠ¸ ì „ì²˜ë¦¬ íŒŒì´í”„ë¼ì¸
# Day 4: ì „ì²˜ë¦¬ + TOP 3 ëª¨ë¸ ë¹„êµ (ëª¨ë“ˆí™” ì ìš© ë²„ì „)

print("âœ… ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸ ì™„ë£Œ!")
print(f"scikit-learn ë²„ì „: {__import__('sklearn').__version__}")

# =============================================================================
# 0. í™˜ê²½ ì„¤ì • (ëª¨ë“ˆí™”ëœ í•¨ìˆ˜ ì‚¬ìš©)
# =============================================================================

# ë¨¸ì‹ ëŸ¬ë‹ ë¼ì´ë¸ŒëŸ¬ë¦¬
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.pipeline import Pipeline

from pathlib import Path
import sys
sys.path.append(r'D:\dataAnalystBook\my_workspace\movie_review') # project_root ê²½ë¡œ
from src.utils.setup import setup_notebook_environment

"""
setup_notebook_environment() í•¨ìˆ˜ê°€ ìˆ˜í–‰í•˜ëŠ” ì‘ì—…:
1. í”„ë¡œì íŠ¸ ë£¨íŠ¸ ìë™ íƒì§€ ë° sys.path ì¶”ê°€
2. í•œê¸€ í°íŠ¸ ì„¤ì • (matplotlibì—ì„œ í•œê¸€ í‘œì‹œ)
3. pandas í‘œì‹œ ì˜µì…˜ ì„¤ì • (ìµœëŒ€ í–‰/ì—´ ìˆ˜ ì¡°ì •)
4. seaborn ìŠ¤íƒ€ì¼ ë° ìƒ‰ìƒ íŒ”ë ˆíŠ¸ ì„¤ì •
5. ê²½ê³  ë©”ì‹œì§€ ìˆ¨ê¹€ ì„¤ì •
6. ì£¼ìš” íŒ¨í‚¤ì§€ ë²„ì „ í™•ì¸ ë° ì¶œë ¥
â†’ ê²°ê³¼: ëª¨ë“  ë…¸íŠ¸ë¶ì—ì„œ ì¼ê´€ëœ í™˜ê²½ êµ¬ì„±
"""
setup_notebook_environment()

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
# =============================================================================
# 1. ë°ì´í„° ë¡œë”© (ëª¨ë“ˆí™”ëœ í•¨ìˆ˜ ì‚¬ìš©)
# =============================================================================

from src.data.data_loader import load_nsmc_data

"""
load_nsmc_data() í•¨ìˆ˜ê°€ ìˆ˜í–‰í•˜ëŠ” ì‘ì—…:
1. ë°ì´í„° ë””ë ‰í† ë¦¬ ìë™ íƒì§€ (data/raw/ í´ë” ì°¾ê¸°)
2. ë‹¤ì¤‘ ì¸ì½”ë”© ì‹œë„ (utf-8, cp949, euc-kr)
3. ë°ì´í„° ê²€ì¦ (ì»¬ëŸ¼ëª…, ë°ì´í„° í¬ê¸°, ë¼ë²¨ê°’ í™•ì¸)
4. ê²°ì¸¡ê°’ ë° ë°ì´í„° í’ˆì§ˆ ì²´í¬
5. ìƒì„¸í•œ ë°ì´í„° ìš”ì•½ ì •ë³´ ì¶œë ¥
â†’ ê²°ê³¼: ì•ˆì „í•˜ê³  ê²€ì¦ëœ NSMC ë°ì´í„° ë¡œë”© ì™„ë£Œ
"""
train_df, test_df = load_nsmc_data()

print(train_df.head())

# ê¸°ë³¸ ì •ë³´ í™•ì¸
print("\nğŸ“‹ ë°ì´í„° ê¸°ë³¸ ì •ë³´:")
print(f"í›ˆë ¨ ë°ì´í„°: {train_df.shape}")
print(f"í…ŒìŠ¤íŠ¸ ë°ì´í„°: {test_df.shape}")
print(f"ì»¬ëŸ¼: {train_df.columns.tolist()}")

# =============================================================================
# 2. í•œêµ­ì–´ í…ìŠ¤íŠ¸ ì „ì²˜ë¦¬ (ëª¨ë“ˆí™”ëœ í•¨ìˆ˜ ì‚¬ìš©)
# =============================================================================

from src.data.text_preprocessor import preprocess_text_data

"""
preprocess_text_data() í•¨ìˆ˜ê°€ ìˆ˜í–‰í•˜ëŠ” ì‘ì—…:
1. í…ìŠ¤íŠ¸ ì •ì œ (clean_korean_text í•¨ìˆ˜ ë‚´ë¶€ í˜¸ì¶œ):
   - ë°˜ë³µ ë¬¸ì ì •ê·œí™” (ã…‹ã…‹ã…‹ â†’ ã…‹ã…‹, !!! â†’ !!)
   - ì˜ë¯¸ì—†ëŠ” íŠ¹ìˆ˜ë¬¸ì ì œê±° (ê°ì • í‘œí˜„ì€ ë³´ì¡´)
   - ì—¬ëŸ¬ ê³µë°±ì„ í•˜ë‚˜ë¡œ í†µí•©
   - ì•ë’¤ ê³µë°± ì œê±°
2. ë¹ˆ í…ìŠ¤íŠ¸ ì œê±° ë° ê°œìˆ˜ ë¦¬í¬íŠ¸
3. í…ìŠ¤íŠ¸ ê¸¸ì´ ê³„ì‚° ë° í†µê³„ ì¶œë ¥
4. ì§„í–‰ ìƒí™© í‘œì‹œ (tqdm í”„ë¡œê·¸ë ˆìŠ¤ ë°”)
â†’ ê²°ê³¼: 'cleaned_text', 'text_length' ì»¬ëŸ¼ì´ ì¶”ê°€ëœ ì •ì œëœ ë°ì´í„°í”„ë ˆì„
"""

print("\n" + "="*50)
print("ğŸ”§ ë°ì´í„° ì „ì²˜ë¦¬ ì‹œì‘")
print("="*50)

# í›ˆë ¨ ë°ì´í„° ì „ì²˜ë¦¬
print("\nğŸ“Š í›ˆë ¨ ë°ì´í„° ì „ì²˜ë¦¬...")
train_processed = preprocess_text_data(train_df)

# í…ŒìŠ¤íŠ¸ ë°ì´í„° ì „ì²˜ë¦¬  
print("\nğŸ“Š í…ŒìŠ¤íŠ¸ ë°ì´í„° ì „ì²˜ë¦¬...")
test_processed = preprocess_text_data(test_df)

# ì „ì²˜ë¦¬ ê²°ê³¼ í™•ì¸
print("\nâœ… ì „ì²˜ë¦¬ ì™„ë£Œ!")
print(f"í›ˆë ¨ ë°ì´í„°: {len(train_processed):,}ê°œ")
print(f"í…ŒìŠ¤íŠ¸ ë°ì´í„°: {len(test_processed):,}ê°œ")

# ì „ì²˜ë¦¬ ì˜ˆì‹œ í™•ì¸
print("\nğŸ“ ì „ì²˜ë¦¬ ì˜ˆì‹œ:")
for i in range(3):
    original = train_df.iloc[i]['document']
    cleaned = train_processed.iloc[i]['cleaned_text']
    print(f"\nì˜ˆì‹œ {i+1}:")
    print(f"ì›ë³¸: {original}")
    print(f"ì •ì œ: {cleaned}")

# =============================================================================
# 3. ë°ì´í„° ë¶„í•  ë° TF-IDF ë²¡í„°í™”
# =============================================================================

print("\n" + "="*50)
print("ğŸ”€ ë°ì´í„° ë¶„í•  ë° ë²¡í„°í™”")
print("="*50)

# í…ìŠ¤íŠ¸ì™€ ë¼ë²¨ ë¶„ë¦¬
X_train_text = train_processed['cleaned_text'].values
y_train = train_processed['label'].values
X_test_text = test_processed['cleaned_text'].values  
y_test = test_processed['label'].values

from sklearn.model_selection import train_test_split

"""
train_test_split() í•¨ìˆ˜ê°€ ìˆ˜í–‰í•˜ëŠ” ì‘ì—…:
1. í›ˆë ¨ ë°ì´í„°ë¥¼ train/validationìœ¼ë¡œ 8:2 ë¶„í• 
2. stratify=y_trainìœ¼ë¡œ í´ë˜ìŠ¤ ë¹„ìœ¨ ìœ ì§€
3. random_state=42ë¡œ ì¬í˜„ ê°€ëŠ¥í•œ ë¶„í• 
â†’ ê²°ê³¼: ê· í˜•ì¡íŒ train/validation ë°ì´í„°ì…‹ ìƒì„±
"""

# í›ˆë ¨ ë°ì´í„°ë¥¼ train/validationìœ¼ë¡œ ë¶„í• 
X_train, X_val, y_train_split, y_val = train_test_split(
    X_train_text, y_train, 
    test_size=0.2, 
    random_state=42, 
    stratify=y_train
)

print(f"ğŸ“Š ë°ì´í„° ë¶„í•  ê²°ê³¼:")
print(f"í›ˆë ¨: {len(X_train):,}ê°œ")
print(f"ê²€ì¦: {len(X_val):,}ê°œ") 
print(f"í…ŒìŠ¤íŠ¸: {len(X_test_text):,}ê°œ")

# TF-IDF ë²¡í„°í™”
print(f"\nğŸ”¢ TF-IDF ë²¡í„°í™” ì¤‘...")

from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

"""
TfidfVectorizer ì„¤ì • ë° ë³€í™˜ ê³¼ì •:
1. TF-IDF ì„¤ì •:
   - max_features=10000: ìµœëŒ€ 10,000ê°œ íŠ¹ì„±ìœ¼ë¡œ ì œí•œ
   - min_df=2: ìµœì†Œ 2ê°œ ë¬¸ì„œì— ë“±ì¥í•˜ëŠ” ë‹¨ì–´ë§Œ ì‚¬ìš©
   - max_df=0.95: 95% ì´ìƒ ë¬¸ì„œì— ë“±ì¥í•˜ëŠ” í”í•œ ë‹¨ì–´ ì œê±°
   - ngram_range=(1,2): ë‹¨ì–´(1-gram)ì™€ 2ë‹¨ì–´ ì¡°í•©(2-gram) ì‚¬ìš©
   - sublinear_tf=True: TF ê°’ì— ë¡œê·¸ ìŠ¤ì¼€ì¼ë§ ì ìš©

2. ë³€í™˜ ê³¼ì •:
   - fit_transform(X_train): í›ˆë ¨ ë°ì´í„°ë¡œ ì–´íœ˜ í•™ìŠµ + ë²¡í„° ë³€í™˜
   - transform(X_val, X_test): í•™ìŠµëœ ì–´íœ˜ë¡œ ê²€ì¦/í…ŒìŠ¤íŠ¸ ë°ì´í„° ë³€í™˜
   
â†’ ê²°ê³¼: í¬ì†Œ í–‰ë ¬ í˜•íƒœì˜ TF-IDF ë²¡í„° (ë¬¸ì„œ x ë‹¨ì–´ íŠ¹ì„±)
"""

# TF-IDF ì„¤ì • (í•œêµ­ì–´ íŠ¹í™”)
from src.features.tfidf_vectorizer import create_tfidf_vectorizer, show_tfidf_stats

# TF-IDF ë²¡í„°ë¼ì´ì € ìƒì„±
tfidf = create_tfidf_vectorizer()

# í›ˆë ¨ ë°ì´í„°ë¡œ TF-IDF í•™ìŠµ ë° ë³€í™˜
X_train_tfidf = tfidf.fit_transform(X_train)
X_val_tfidf = tfidf.transform(X_val)
X_test_tfidf = tfidf.transform(X_test_text)

print(f"âœ… TF-IDF ë²¡í„°í™” ì™„ë£Œ!")
print(f"ğŸ”¢ íŠ¹ì„± ìˆ˜: {X_train_tfidf.shape[1]:,}ê°œ")
print(f"ğŸ“Š í¬ì†Œì„±: {(1 - X_train_tfidf.nnz / np.prod(X_train_tfidf.shape)) * 100:.1f}%")

# ì£¼ìš” íŠ¹ì„± í™•ì¸
feature_names = tfidf.get_feature_names_out()
print(f"\nğŸ“ ì£¼ìš” íŠ¹ì„± ì˜ˆì‹œ (ì²˜ìŒ 20ê°œ):")
print(feature_names[:20])

print(f"\nğŸ¯ ì „ì²˜ë¦¬ ë‹¨ê³„ ì™„ë£Œ! ì´ì œ ëª¨ë¸ í•™ìŠµì„ ì‹œì‘í•©ë‹ˆë‹¤.")
print(f"â° ì†Œìš” ì‹œê°„: ì•½ 5-10ë¶„ ì˜ˆìƒ")

# =============================================================================
# 4. ëª¨ë“ˆí™” íš¨ê³¼ ìš”ì•½
# =============================================================================

print(f"\nğŸ“‹ ëª¨ë“ˆí™”ëœ í•¨ìˆ˜ë“¤ì´ ìˆ˜í–‰í•œ ì‘ì—… ìš”ì•½:")
print(f"âœ… setup_notebook_environment(): í™˜ê²½ ì„¤ì • ìë™í™”")
print(f"   â†’ í”„ë¡œì íŠ¸ ê²½ë¡œ, í°íŠ¸, pandas ì˜µì…˜, íŒ¨í‚¤ì§€ ë²„ì „ í™•ì¸")
print(f"âœ… load_nsmc_data(): ì•ˆì „í•œ ë°ì´í„° ë¡œë”©")
print(f"   â†’ ìë™ ê²½ë¡œ íƒì§€, ì¸ì½”ë”© ì²˜ë¦¬, ë°ì´í„° ê²€ì¦")
print(f"âœ… preprocess_text_data(): í•œêµ­ì–´ íŠ¹í™” ì „ì²˜ë¦¬")
print(f"   â†’ í…ìŠ¤íŠ¸ ì •ì œ, ë¹ˆ ê°’ ì œê±°, ê¸¸ì´ ê³„ì‚°, í†µê³„ ì¶œë ¥")

print(f"\nğŸ¯ ìµœì¢… ì¤€ë¹„ ì™„ë£Œ:")
print(f"  - ì •ì œëœ í…ìŠ¤íŠ¸ ë°ì´í„°: {len(train_processed):,} + {len(test_processed):,}ê°œ")
print(f"  - TF-IDF ë²¡í„°: {X_train_tfidf.shape} (í›ˆë ¨)")
print(f"  - ê²€ì¦/í…ŒìŠ¤íŠ¸ ë²¡í„°: {X_val_tfidf.shape}, {X_test_tfidf.shape}")
print(f"  - ë‹¤ìŒ ë‹¨ê³„: ë¨¸ì‹ ëŸ¬ë‹ ëª¨ë¸ í•™ìŠµ (ë‚˜ì´ë¸Œë² ì´ì¦ˆ, ë¡œì§€ìŠ¤í‹±íšŒê·€, SVM)")
```

```python
# =============================================================================
# ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ (Naive Bayes) ìƒì„¸ ë¶„ì„
# =============================================================================

import time
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

print("ğŸ§  ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ (Naive Bayes) ìƒì„¸ ë¶„ì„ ì‹œì‘!")
print("="*60)

# =============================================================================
# 1. ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ ëª¨ë¸ ì •ì˜ ë° ì„¤ëª…
# =============================================================================

print("ğŸ“š ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ ëª¨ë¸ ì„¤ì •:")
print("â–ªï¸ ì•Œê³ ë¦¬ì¦˜: MultinomialNB (ë‹¤í•­ ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ)")
print("â–ªï¸ íŠ¹ì§•: í…ìŠ¤íŠ¸ ë¶„ë¥˜ì— íŠ¹í™”ëœ ë²„ì „")
print("â–ªï¸ ê°€ì •: ë‹¨ì–´ë“¤ì´ ì„œë¡œ ë…ë¦½ì ")
print("â–ªï¸ ì¥ì : ë¹ ë¥¸ í•™ìŠµ, ë†’ì€ ì„±ëŠ¥, í•´ì„ ê°€ëŠ¥")

# ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ ëª¨ë¸ ìƒì„±
nb_model = MultinomialNB(alpha=1.0)  # alpha: ë¼í”Œë¼ìŠ¤ ìŠ¤ë¬´ë”© (ê¸°ë³¸ê°’ 1.0)

print(f"\nâš™ï¸ í•˜ì´í¼íŒŒë¼ë¯¸í„°:")
print(f"â–ªï¸ alpha (ìŠ¤ë¬´ë”©): {nb_model.alpha}")
print(f"   â†’ 0ì´ ë‚˜ì˜¤ì§€ ì•ŠëŠ” ë‹¨ì–´ì— ëŒ€í•œ ì²˜ë¦¬ (0í™•ë¥  ë°©ì§€)")

# =============================================================================
# 2. êµì°¨ ê²€ì¦ìœ¼ë¡œ ì•ˆì •ì  ì„±ëŠ¥ ì¸¡ì •
# =============================================================================

print(f"\nğŸ”¬ 5-fold êµì°¨ ê²€ì¦ ì‹¤í–‰...")
print(f"ğŸ“Š ë°ì´í„°: {X_train_tfidf.shape[0]:,}ê°œ ìƒ˜í”Œ Ã— {X_train_tfidf.shape[1]:,}ê°œ íŠ¹ì„±")

start_time = time.time()

# 5-fold êµì°¨ ê²€ì¦ ì‹¤í–‰
cv_scores = cross_val_score(
    nb_model, 
    X_train_tfidf, 
    y_train_split, 
    cv=5, 
    scoring='accuracy'
)

cv_time = time.time() - start_time

print(f"âœ… êµì°¨ ê²€ì¦ ì™„ë£Œ!")
print(f"â° ì†Œìš” ì‹œê°„: {cv_time:.2f}ì´ˆ")
print(f"\nğŸ“Š êµì°¨ ê²€ì¦ ê²°ê³¼:")
print(f"â–ªï¸ í‰ê·  ì •í™•ë„: {cv_scores.mean():.4f}")
print(f"â–ªï¸ í‘œì¤€í¸ì°¨: {cv_scores.std():.4f}")
print(f"â–ªï¸ 95% ì‹ ë¢°êµ¬ê°„: {cv_scores.mean():.4f} Â± {cv_scores.std()*2:.4f}")

print(f"\nğŸ“ˆ ê° foldë³„ ì„±ëŠ¥:")
for i, score in enumerate(cv_scores, 1):
    print(f"   Fold {i}: {score:.4f}")

# =============================================================================
# 3. ì „ì²´ í›ˆë ¨ ë°ì´í„°ë¡œ ìµœì¢… í•™ìŠµ
# =============================================================================

print(f"\nğŸ“ ì „ì²´ í›ˆë ¨ ë°ì´í„°ë¡œ ìµœì¢… í•™ìŠµ...")

start_time = time.time()
nb_model.fit(X_train_tfidf, y_train_split)
train_time = time.time() - start_time

print(f"âœ… í•™ìŠµ ì™„ë£Œ! ì†Œìš” ì‹œê°„: {train_time:.3f}ì´ˆ")

# ê²€ì¦ ë°ì´í„°ë¡œ ì˜ˆì¸¡
val_predictions = nb_model.predict(X_val_tfidf)
val_probabilities = nb_model.predict_proba(X_val_tfidf)

# ì„±ëŠ¥ ê³„ì‚°
val_accuracy = accuracy_score(y_val, val_predictions)

print(f"\nğŸ“Š ê²€ì¦ ë°ì´í„° ì„±ëŠ¥:")
print(f"â–ªï¸ ì •í™•ë„: {val_accuracy:.4f}")

# ìƒì„¸ ì„±ëŠ¥ ë¶„ì„
report = classification_report(y_val, val_predictions, 
                             target_names=['ë¶€ì •', 'ê¸ì •'], 
                             output_dict=True)

print(f"â–ªï¸ ì •ë°€ë„ (ê¸ì •): {report['ê¸ì •']['precision']:.4f}")
print(f"â–ªï¸ ì¬í˜„ìœ¨ (ê¸ì •): {report['ê¸ì •']['recall']:.4f}")
print(f"â–ªï¸ F1-Score (ê¸ì •): {report['ê¸ì •']['f1-score']:.4f}")

print(f"â–ªï¸ ì •ë°€ë„ (ë¶€ì •): {report['ë¶€ì •']['precision']:.4f}")
print(f"â–ªï¸ ì¬í˜„ìœ¨ (ë¶€ì •): {report['ë¶€ì •']['recall']:.4f}")
print(f"â–ªï¸ F1-Score (ë¶€ì •): {report['ë¶€ì •']['f1-score']:.4f}")

# =============================================================================
# 4. ë‚˜ì´ë¸Œ ë² ì´ì¦ˆì˜ í•µì‹¬: ë‹¨ì–´ë³„ í™•ë¥  ë¶„ì„
# =============================================================================

print(f"\nğŸ” ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ í•µì‹¬ ë¶„ì„: ë‹¨ì–´ë³„ í™•ë¥ ")
print("="*60)

# íŠ¹ì„±ëª… ê°€ì ¸ì˜¤ê¸°
feature_names = tfidf.get_feature_names_out()

# ê° í´ë˜ìŠ¤ë³„ ë¡œê·¸ í™•ë¥  (ë¶€ì •=0, ê¸ì •=1)
log_prob_neg = nb_model.feature_log_prob_[0]  # ë¶€ì • í´ë˜ìŠ¤
log_prob_pos = nb_model.feature_log_prob_[1]  # ê¸ì • í´ë˜ìŠ¤

# ë¡œê·¸ í™•ë¥  ì°¨ì´ ê³„ì‚° (ê¸ì • - ë¶€ì •)
log_prob_diff = log_prob_pos - log_prob_neg

print(f"ğŸ“ˆ ê°€ì¥ ê°•í•œ ê¸ì • ì‹ í˜¸ ë‹¨ì–´ë“¤ (TOP 15):")
top_positive_idx = np.argsort(log_prob_diff)[-15:]
for i, idx in enumerate(top_positive_idx[::-1], 1):
    word = feature_names[idx]
    diff = log_prob_diff[idx]
    pos_prob = np.exp(log_prob_pos[idx])
    neg_prob = np.exp(log_prob_neg[idx])
    print(f"   {i:2d}. {word:15} | ì°¨ì´: {diff:6.3f} | ê¸ì •í™•ë¥ : {pos_prob:.6f} | ë¶€ì •í™•ë¥ : {neg_prob:.6f}")

print(f"\nğŸ“‰ ê°€ì¥ ê°•í•œ ë¶€ì • ì‹ í˜¸ ë‹¨ì–´ë“¤ (TOP 15):")
top_negative_idx = np.argsort(log_prob_diff)[:15]
for i, idx in enumerate(top_negative_idx, 1):
    word = feature_names[idx]
    diff = log_prob_diff[idx]
    pos_prob = np.exp(log_prob_pos[idx])
    neg_prob = np.exp(log_prob_neg[idx])
    print(f"   {i:2d}. {word:15} | ì°¨ì´: {diff:6.3f} | ê¸ì •í™•ë¥ : {pos_prob:.6f} | ë¶€ì •í™•ë¥ : {neg_prob:.6f}")

# =============================================================================
# 5. ì‹¤ì œ ì˜ˆì¸¡ ê³¼ì • ì‹œì—°
# =============================================================================

print(f"\nğŸ¯ ì‹¤ì œ ì˜ˆì¸¡ ê³¼ì • ì‹œì—°:")
print("="*60)

# ê²€ì¦ ë°ì´í„°ì—ì„œ ëª‡ ê°œ ìƒ˜í”Œ ì„ íƒ
sample_indices = [0, 100, 500, 1000, 2000]

for i, idx in enumerate(sample_indices):
    actual_text = X_val[idx]
    actual_label = y_val[idx]
    predicted_label = val_predictions[idx]
    predicted_proba = val_probabilities[idx]
    
    print(f"\nì˜ˆì‹œ {i+1}:")
    print(f"ì›ë³¸ í…ìŠ¤íŠ¸: {actual_text}")
    print(f"ì‹¤ì œ ë¼ë²¨: {'ê¸ì •' if actual_label == 1 else 'ë¶€ì •'}")
    print(f"ì˜ˆì¸¡ ë¼ë²¨: {'ê¸ì •' if predicted_label == 1 else 'ë¶€ì •'}")
    print(f"ì˜ˆì¸¡ í™•ë¥ : ë¶€ì • {predicted_proba[0]:.3f}, ê¸ì • {predicted_proba[1]:.3f}")
    
    # ì˜ˆì¸¡ì´ ë§ì•˜ëŠ”ì§€ í™•ì¸
    if actual_label == predicted_label:
        print("âœ… ì •ë‹µ!")
    else:
        print("âŒ ì˜¤ë‹µ!")

# =============================================================================
# 6. í˜¼ë™ í–‰ë ¬ ì‹œê°í™”
# =============================================================================

print(f"\nğŸ“Š ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ í˜¼ë™ í–‰ë ¬:")

# í˜¼ë™ í–‰ë ¬ ê³„ì‚°
cm = confusion_matrix(y_val, val_predictions)

# ì‹œê°í™”
plt.figure(figsize=(8, 6))
sns.heatmap(cm, 
            annot=True, 
            fmt='d', 
            cmap='Blues',
            xticklabels=['ë¶€ì •', 'ê¸ì •'],
            yticklabels=['ë¶€ì •', 'ê¸ì •'],
            cbar_kws={'label': 'ìƒ˜í”Œ ìˆ˜'})

plt.title(f'ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ í˜¼ë™ í–‰ë ¬\nì •í™•ë„: {val_accuracy:.4f}', 
          fontsize=14, fontweight='bold')
plt.xlabel('ì˜ˆì¸¡ ë¼ë²¨', fontsize=12)
plt.ylabel('ì‹¤ì œ ë¼ë²¨', fontsize=12)

# ê° ì…€ì— ë°±ë¶„ìœ¨ ì¶”ê°€
total = cm.sum()
for i in range(2):
    for j in range(2):
        percentage = cm[i, j] / total * 100
        plt.text(j + 0.5, i + 0.7, f'({percentage:.1f}%)', 
                ha='center', va='center', fontsize=10, color='gray')

plt.tight_layout()
plt.show()

# =============================================================================
# 7. ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ ì„±ëŠ¥ ìš”ì•½
# =============================================================================

print(f"\nğŸ¯ ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ ìµœì¢… ì„±ëŠ¥ ìš”ì•½:")
print("="*60)

print(f"ğŸ“Š í•µì‹¬ ì„±ëŠ¥ ì§€í‘œ:")
print(f"â–ªï¸ êµì°¨ ê²€ì¦ ì •í™•ë„: {cv_scores.mean():.4f} Â± {cv_scores.std():.4f}")
print(f"â–ªï¸ ê²€ì¦ ë°ì´í„° ì •í™•ë„: {val_accuracy:.4f}")
print(f"â–ªï¸ ì „ì²´ F1-Score: {report['macro avg']['f1-score']:.4f}")

print(f"\nâš¡ íš¨ìœ¨ì„± ì§€í‘œ:")
print(f"â–ªï¸ êµì°¨ ê²€ì¦ ì‹œê°„: {cv_time:.2f}ì´ˆ")
print(f"â–ªï¸ ìµœì¢… í•™ìŠµ ì‹œê°„: {train_time:.3f}ì´ˆ")
print(f"â–ªï¸ ì´ˆë‹¹ ì²˜ë¦¬ ìƒ˜í”Œ: {len(X_train)/train_time:,.0f}ê°œ")

print(f"\nğŸ¯ ëª©í‘œ ë‹¬ì„± ì—¬ë¶€:")
target_75 = 0.75
target_80 = 0.80

if val_accuracy >= target_80:
    print(f"ğŸš€ ìš°ìˆ˜! 80% ëª©í‘œ ë‹¬ì„±: {val_accuracy:.4f} >= {target_80:.4f}")
elif val_accuracy >= target_75:
    print(f"âœ… ì„±ê³µ! 75% ëª©í‘œ ë‹¬ì„±: {val_accuracy:.4f} >= {target_75:.4f}")
else:
    print(f"âš ï¸  ëª©í‘œ ë¯¸ë‹¬ì„±: {val_accuracy:.4f} < {target_75:.4f}")

print(f"\nğŸ’¡ ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ íŠ¹ì§• ìš”ì•½:")
print(f"âœ… ì¥ì : ë§¤ìš° ë¹ ë¥¸ í•™ìŠµ ({train_time:.3f}ì´ˆ)")
print(f"âœ… ì¥ì : ì™„ë²½í•œ í•´ì„ ê°€ëŠ¥ì„± (ë‹¨ì–´ë³„ í™•ë¥  í™•ì¸)")
print(f"âœ… ì¥ì : í…ìŠ¤íŠ¸ ë¶„ë¥˜ì— íŠ¹í™”ëœ ì•Œê³ ë¦¬ì¦˜")
print(f"âœ… ì¥ì : ë©”ëª¨ë¦¬ íš¨ìœ¨ì ")
print(f"âš ï¸  ë‹¨ì : ë‹¨ì–´ ë…ë¦½ì„± ê°€ì •ì˜ í•œê³„")
print(f"âš ï¸  ë‹¨ì : ë³µì¡í•œ ë‹¨ì–´ ì¡°í•© íŒ¨í„´ ë†“ì¹  ìˆ˜ ìˆìŒ")

print(f"\nğŸ ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ ë¶„ì„ ì™„ë£Œ!")
print(f"â° ì™„ë£Œ ì‹œê°„: {datetime.now().strftime('%H:%M:%S')}")

# ê²°ê³¼ë¥¼ ë‹¤ìŒ ë‹¨ê³„ë¥¼ ìœ„í•´ ì €ì¥
nb_results = {
    'model': nb_model,
    'cv_scores': cv_scores,
    'cv_mean': cv_scores.mean(),
    'cv_std': cv_scores.std(),
    'val_accuracy': val_accuracy,
    'train_time': train_time,
    'cv_time': cv_time,
    'predictions': val_predictions,
    'probabilities': val_probabilities,
    'classification_report': report
}

print(f"ğŸ’¾ ê²°ê³¼ ì €ì¥ ì™„ë£Œ: nb_results ë³€ìˆ˜ì— ì €ì¥ë¨")

# =============================================================================
# 8. ëª¨ë¸ ì €ì¥
# =============================================================================

import joblib

print(f"\nğŸ’¾ ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ ëª¨ë¸ ì €ì¥...")

# ëª¨ë¸ ì €ì¥
joblib.dump(nb_model, '../../models/trained/naive_bayes_v1.pkl')

# TF-IDF ë²¡í„°ë¼ì´ì €ë„ í•¨ê»˜ ì €ì¥ (Day 4 ê¸°ë³¸ ì„¤ì •)
joblib.dump(tfidf, '../../models/trained/tfidf_vectorizer_v1.pkl')

print(f"âœ… ì €ì¥ ì™„ë£Œ:")
print(f"   â–ªï¸ ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ: models/trained/naive_bayes_v1.pkl")
print(f"   â–ªï¸ TF-IDF ë²¡í„°ë¼ì´ì €: models/trained/tfidf_vectorizer_v1.pkl")


# =============================================================================
# 9. ì˜ˆì¸¡ ê²°ê³¼ ì €ì¥
# =============================================================================

print(f"\nğŸ’¾ ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ ì˜ˆì¸¡ ê²°ê³¼ ì €ì¥...")

# ì˜ˆì¸¡ ê²°ê³¼ DataFrame ìƒì„±
import pandas as pd
import os

# results í´ë” ìƒì„± (ì—†ìœ¼ë©´)
os.makedirs('../../results/predictions', exist_ok=True)

predictions_df = pd.DataFrame({
    'actual_label': y_val,
    'predicted_label': val_predictions,
    'predicted_prob_negative': val_probabilities[:, 0],
    'predicted_prob_positive': val_probabilities[:, 1],
    'review_text': X_val,
    'correct_prediction': (y_val == val_predictions).astype(int),
    'confidence': np.max(val_probabilities, axis=1)
})

# í•œê¸€ ì¸ì½”ë”©ìœ¼ë¡œ CSV ì €ì¥
predictions_df.to_csv(
    '../../results/predictions/naive_bayes_predictions.csv', 
    index=False, 
    encoding='utf-8-sig',  # Excelì—ì„œë„ í•œê¸€ì´ ì œëŒ€ë¡œ ë³´ì´ë„ë¡
    quoting=1  # ëª¨ë“  ë¬¸ìì—´ì„ ë”°ì˜´í‘œë¡œ ê°ì‹¸ê¸°
)

print(f"âœ… ì˜ˆì¸¡ ê²°ê³¼ ì €ì¥ ì™„ë£Œ:")
print(f"   â–ªï¸ íŒŒì¼: results/predictions/naive_bayes_predictions.csv")
print(f"   â–ªï¸ ì´ {len(predictions_df):,}ê°œ ì˜ˆì¸¡ ê²°ê³¼")

# ì„±ëŠ¥ ì§€í‘œë„ JSONìœ¼ë¡œ ì €ì¥
os.makedirs('../../results/metrics', exist_ok=True)

nb_metrics = {
    'model_name': 'naive_bayes_v1',
    'model_type': 'MultinomialNB',
    'evaluation_date': datetime.now().isoformat(),
    'validation_performance': {
        'accuracy': float(val_accuracy),
        'precision_positive': float(report['ê¸ì •']['precision']),
        'recall_positive': float(report['ê¸ì •']['recall']),
        'f1_positive': float(report['ê¸ì •']['f1-score']),
        'precision_negative': float(report['ë¶€ì •']['precision']),
        'recall_negative': float(report['ë¶€ì •']['recall']),
        'f1_negative': float(report['ë¶€ì •']['f1-score']),
        'macro_avg_f1': float(report['macro avg']['f1-score'])
    },
    'cross_validation': {
        'cv_mean': float(cv_scores.mean()),
        'cv_std': float(cv_scores.std()),
        'cv_scores': [float(score) for score in cv_scores]
    },
    'training_info': {
        'train_time': float(train_time),
        'cv_time': float(cv_time),
        'alpha': float(nb_model.alpha)
    }
}

import json
with open('../../results/metrics/naive_bayes_metrics.json', 'w', encoding='utf-8') as f:
    json.dump(nb_metrics, f, indent=2, ensure_ascii=False)

print(f"   â–ªï¸ ì„±ëŠ¥ ì§€í‘œ: results/metrics/naive_bayes_metrics.json")
```

```python
# =============================================================================
# 1ë‹¨ê³„: Soft Voting ì•™ìƒë¸” ëª¨ë¸ êµ¬í˜„
# =============================================================================

import time
from datetime import datetime
import numpy as np
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

print("ğŸ—³ï¸ 1ë‹¨ê³„: Soft Voting ì•™ìƒë¸” ëª¨ë¸ êµ¬í˜„")
print("="*60)

# =============================================================================
# 1. ì•™ìƒë¸” ëª¨ë¸ ì •ì˜ ë° ì„¤ëª…
# =============================================================================

print("ğŸ“š Soft Voting ì•™ìƒë¸” ì„¤ëª…:")
print("â–ªï¸ ëª©ì : ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ + ë¡œì§€ìŠ¤í‹± íšŒê·€ ì¡°í•©")
print("â–ªï¸ ë°©ë²•: ë‘ ëª¨ë¸ì˜ ì˜ˆì¸¡ í™•ë¥ ì„ í‰ê· í•˜ì—¬ ê²°í•©")
print("â–ªï¸ ì¥ì : í™•ì‹ ë„ë¥¼ ê³ ë ¤í•œ ë” ì •êµí•œ íŒë‹¨")
print("â–ªï¸ ëª©í‘œ: 79.25% â†’ 80%+ ì •í™•ë„ í–¥ìƒ")

# ê¸°ì¡´ ëª¨ë¸ë“¤ ì„±ëŠ¥ í™•ì¸
print(f"\nğŸ“Š ê¸°ì¡´ ê°œë³„ ëª¨ë¸ ì„±ëŠ¥:")
print(f"â–ªï¸ ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ: {nb_results['val_accuracy']:.4f}")
print(f"â–ªï¸ ë¡œì§€ìŠ¤í‹± íšŒê·€: {lr_results['val_accuracy']:.4f}")
print(f"â–ªï¸ ì„±ëŠ¥ ì°¨ì´: {abs(nb_results['val_accuracy'] - lr_results['val_accuracy']):.4f}p")

# =============================================================================
# 2. Soft Voting ì•™ìƒë¸” ìƒì„±
# =============================================================================

print(f"\nğŸ”§ Soft Voting ì•™ìƒë¸” ìƒì„±...")

# ì•™ìƒë¸” ëª¨ë¸ ì •ì˜
soft_voting_clf = VotingClassifier(
    estimators=[
        ('naive_bayes', MultinomialNB(alpha=1.0)),
        ('logistic_regression', LogisticRegression(
            C=1.0, 
            max_iter=1000, 
            solver='liblinear', 
            random_state=42
        ))
    ],
    voting='soft',  # ì†Œí”„íŠ¸ ë³´íŒ… ì‚¬ìš©
    weights=None    # ë™ë“±í•œ ê°€ì¤‘ì¹˜ (1:1)
)

print("âœ… Soft Voting ì•™ìƒë¸” ëª¨ë¸ ìƒì„± ì™„ë£Œ")
print("â–ªï¸ êµ¬ì„±: ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ + ë¡œì§€ìŠ¤í‹± íšŒê·€")
print("â–ªï¸ ë³´íŒ…: Soft (í™•ë¥  í‰ê· )")
print("â–ªï¸ ê°€ì¤‘ì¹˜: ë™ë“± (1:1)")

# =============================================================================
# 3. êµì°¨ ê²€ì¦ìœ¼ë¡œ ì„±ëŠ¥ ì¸¡ì •
# =============================================================================

print(f"\nğŸ”¬ Soft Voting êµì°¨ ê²€ì¦ ì‹¤í–‰...")
print(f"ğŸ“Š ë°ì´í„°: {X_train_tfidf.shape[0]:,}ê°œ ìƒ˜í”Œ")

start_time = time.time()

# 5-fold êµì°¨ ê²€ì¦
cv_scores_ensemble = cross_val_score(
    soft_voting_clf, 
    X_train_tfidf, 
    y_train_split, 
    cv=5, 
    scoring='accuracy'
)

cv_time_ensemble = time.time() - start_time

print(f"âœ… êµì°¨ ê²€ì¦ ì™„ë£Œ! ì†Œìš” ì‹œê°„: {cv_time_ensemble:.2f}ì´ˆ")

print(f"\nğŸ“Š Soft Voting êµì°¨ ê²€ì¦ ê²°ê³¼:")
print(f"â–ªï¸ í‰ê·  ì •í™•ë„: {cv_scores_ensemble.mean():.4f}")
print(f"â–ªï¸ í‘œì¤€í¸ì°¨: {cv_scores_ensemble.std():.4f}")
print(f"â–ªï¸ 95% ì‹ ë¢°êµ¬ê°„: {cv_scores_ensemble.mean():.4f} Â± {cv_scores_ensemble.std()*2:.4f}")

print(f"\nğŸ“ˆ ê° foldë³„ ì„±ëŠ¥:")
for i, score in enumerate(cv_scores_ensemble, 1):
    print(f"   Fold {i}: {score:.4f}")

# =============================================================================
# 4. ê°œë³„ ëª¨ë¸ vs ì•™ìƒë¸” ì„±ëŠ¥ ë¹„êµ
# =============================================================================

print(f"\nğŸ†š êµì°¨ ê²€ì¦ ì„±ëŠ¥ ë¹„êµ:")
print("="*60)

# ì„±ëŠ¥ ê°œì„  ê³„ì‚°
nb_cv_mean = nb_results['cv_mean']
lr_cv_mean = lr_results['cv_mean']
ensemble_cv_mean = cv_scores_ensemble.mean()

nb_improvement = ensemble_cv_mean - nb_cv_mean
lr_improvement = ensemble_cv_mean - lr_cv_mean

print(f"ğŸ“Š ì„±ëŠ¥ ìˆœìœ„:")
models_cv_comparison = [
    ("Soft Voting ì•™ìƒë¸”", ensemble_cv_mean, cv_time_ensemble),
    ("ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ", nb_cv_mean, nb_results['cv_time']),
    ("ë¡œì§€ìŠ¤í‹± íšŒê·€", lr_cv_mean, lr_results['cv_time'])
]

# ì„±ëŠ¥ìˆœ ì •ë ¬
models_cv_sorted = sorted(models_cv_comparison, key=lambda x: x[1], reverse=True)

for i, (name, score, time_taken) in enumerate(models_cv_sorted, 1):
    if i == 1:
        print(f"ğŸ¥‡ {i}ìœ„. {name}: {score:.4f} ({time_taken:.2f}ì´ˆ)")
    elif i == 2:
        print(f"ğŸ¥ˆ {i}ìœ„. {name}: {score:.4f} ({time_taken:.2f}ì´ˆ)")
    else:
        print(f"ğŸ¥‰ {i}ìœ„. {name}: {score:.4f} ({time_taken:.2f}ì´ˆ)")

print(f"\nğŸ“ˆ ì•™ìƒë¸” ê°œì„  íš¨ê³¼:")
print(f"â–ªï¸ vs ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ: {nb_improvement:+.4f}p ({nb_improvement*100:+.2f}%p)")
print(f"â–ªï¸ vs ë¡œì§€ìŠ¤í‹± íšŒê·€: {lr_improvement:+.4f}p ({lr_improvement*100:+.2f}%p)")

if ensemble_cv_mean > max(nb_cv_mean, lr_cv_mean):
    print(f"âœ… ì•™ìƒë¸”ì´ ê°œë³„ ëª¨ë¸ë“¤ë³´ë‹¤ ìš°ìˆ˜í•œ ì„±ëŠ¥!")
elif abs(ensemble_cv_mean - max(nb_cv_mean, lr_cv_mean)) < 0.001:
    print(f"ğŸ¤ ì•™ìƒë¸”ì´ ìµœê³  ê°œë³„ ëª¨ë¸ê³¼ ë™ë“±í•œ ì„±ëŠ¥")
else:
    print(f"âš ï¸  ì•™ìƒë¸”ì´ ê°œë³„ ëª¨ë¸ë³´ë‹¤ ë‚®ì€ ì„±ëŠ¥ (ê³¼ì í•© ê°€ëŠ¥ì„±)")

# =============================================================================
# 5. ìµœì¢… í•™ìŠµ ë° ê²€ì¦ ë°ì´í„° í‰ê°€
# =============================================================================

print(f"\nğŸ“ ì „ì²´ í›ˆë ¨ ë°ì´í„°ë¡œ ìµœì¢… í•™ìŠµ...")

start_time = time.time()
soft_voting_clf.fit(X_train_tfidf, y_train_split)
train_time_ensemble = time.time() - start_time

print(f"âœ… í•™ìŠµ ì™„ë£Œ! ì†Œìš” ì‹œê°„: {train_time_ensemble:.3f}ì´ˆ")

# ê²€ì¦ ë°ì´í„°ë¡œ ì˜ˆì¸¡
ensemble_predictions = soft_voting_clf.predict(X_val_tfidf)
ensemble_probabilities = soft_voting_clf.predict_proba(X_val_tfidf)

# ì„±ëŠ¥ ê³„ì‚°
ensemble_accuracy = accuracy_score(y_val, ensemble_predictions)

print(f"\nğŸ“Š ê²€ì¦ ë°ì´í„° ì„±ëŠ¥:")
print(f"â–ªï¸ ì •í™•ë„: {ensemble_accuracy:.4f}")

# ìƒì„¸ ì„±ëŠ¥ ë¶„ì„
ensemble_report = classification_report(y_val, ensemble_predictions, 
                                      target_names=['ë¶€ì •', 'ê¸ì •'], 
                                      output_dict=True)

print(f"â–ªï¸ ì •ë°€ë„ (ê¸ì •): {ensemble_report['ê¸ì •']['precision']:.4f}")
print(f"â–ªï¸ ì¬í˜„ìœ¨ (ê¸ì •): {ensemble_report['ê¸ì •']['recall']:.4f}")
print(f"â–ªï¸ F1-Score (ê¸ì •): {ensemble_report['ê¸ì •']['f1-score']:.4f}")

print(f"â–ªï¸ ì •ë°€ë„ (ë¶€ì •): {ensemble_report['ë¶€ì •']['precision']:.4f}")
print(f"â–ªï¸ ì¬í˜„ìœ¨ (ë¶€ì •): {ensemble_report['ë¶€ì •']['recall']:.4f}")
print(f"â–ªï¸ F1-Score (ë¶€ì •): {ensemble_report['ë¶€ì •']['f1-score']:.4f}")

# =============================================================================
# 6. ìµœì¢… ì„±ëŠ¥ ë¹„êµ (ê²€ì¦ ë°ì´í„°)
# =============================================================================

print(f"\nğŸ†š ìµœì¢… ê²€ì¦ ë°ì´í„° ì„±ëŠ¥ ë¹„êµ:")
print("="*60)

# ê°œë³„ ëª¨ë¸ ì„±ëŠ¥
nb_val_accuracy = nb_results['val_accuracy']
lr_val_accuracy = lr_results['val_accuracy']

# ê°œì„  íš¨ê³¼ ê³„ì‚°
final_nb_improvement = ensemble_accuracy - nb_val_accuracy
final_lr_improvement = ensemble_accuracy - lr_val_accuracy

models_final_comparison = [
    ("Soft Voting ì•™ìƒë¸”", ensemble_accuracy, train_time_ensemble),
    ("ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ", nb_val_accuracy, nb_results['train_time']),
    ("ë¡œì§€ìŠ¤í‹± íšŒê·€", lr_val_accuracy, lr_results['train_time'])
]

# ì„±ëŠ¥ìˆœ ì •ë ¬
models_final_sorted = sorted(models_final_comparison, key=lambda x: x[1], reverse=True)

print(f"ğŸ“Š ìµœì¢… ì„±ëŠ¥ ìˆœìœ„:")
for i, (name, accuracy, time_taken) in enumerate(models_final_sorted, 1):
    if i == 1:
        print(f"ğŸ¥‡ {i}ìœ„. {name}: {accuracy:.4f} (í•™ìŠµ: {time_taken:.3f}ì´ˆ)")
    elif i == 2:
        print(f"ğŸ¥ˆ {i}ìœ„. {name}: {accuracy:.4f} (í•™ìŠµ: {time_taken:.3f}ì´ˆ)")
    else:
        print(f"ğŸ¥‰ {i}ìœ„. {name}: {accuracy:.4f} (í•™ìŠµ: {time_taken:.3f}ì´ˆ)")

print(f"\nğŸ“ˆ ìµœì¢… ì•™ìƒë¸” ê°œì„  íš¨ê³¼:")
print(f"â–ªï¸ vs ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ: {final_nb_improvement:+.4f}p ({final_nb_improvement*100:+.2f}%p)")
print(f"â–ªï¸ vs ë¡œì§€ìŠ¤í‹± íšŒê·€: {final_lr_improvement:+.4f}p ({final_lr_improvement*100:+.2f}%p)")

# =============================================================================
# 7. ê°œë³„ ëª¨ë¸ ì˜ˆì¸¡ê³¼ ì•™ìƒë¸” ì˜ˆì¸¡ ë¹„êµ
# =============================================================================

print(f"\nğŸ” ê°œë³„ ëª¨ë¸ vs ì•™ìƒë¸” ì˜ˆì¸¡ ë¹„êµ (ìƒ˜í”Œ ë¶„ì„):")
print("="*60)

# ê²€ì¦ ë°ì´í„°ì—ì„œ ëª‡ ê°œ ìƒ˜í”Œ ì„ íƒ
sample_indices = [0, 100, 500, 1000, 2000]

for i, idx in enumerate(sample_indices):
    actual_text = X_val[idx]
    actual_label = y_val[idx]
    
    # ê°œë³„ ëª¨ë¸ ì˜ˆì¸¡
    nb_pred = nb_results['predictions'][idx]
    nb_proba = nb_results['probabilities'][idx]
    
    lr_pred = lr_results['predictions'][idx]
    lr_proba = lr_results['probabilities'][idx]
    
    # ì•™ìƒë¸” ì˜ˆì¸¡
    ensemble_pred = ensemble_predictions[idx]
    ensemble_proba = ensemble_probabilities[idx]
    
    print(f"\nì˜ˆì‹œ {i+1}:")
    print(f"í…ìŠ¤íŠ¸: {actual_text}")
    print(f"ì‹¤ì œ: {'ê¸ì •' if actual_label == 1 else 'ë¶€ì •'}")
    
    print(f"NB:       {'ê¸ì •' if nb_pred == 1 else 'ë¶€ì •'} | í™•ë¥ : {nb_proba[1]:.3f}")
    print(f"LR:       {'ê¸ì •' if lr_pred == 1 else 'ë¶€ì •'} | í™•ë¥ : {lr_proba[1]:.3f}")
    print(f"ì•™ìƒë¸”:   {'ê¸ì •' if ensemble_pred == 1 else 'ë¶€ì •'} | í™•ë¥ : {ensemble_proba[1]:.3f}")
    
    # í‰ê·  ê³„ì‚° í™•ì¸
    avg_proba = (nb_proba[1] + lr_proba[1]) / 2
    print(f"í™•ë¥  í‰ê· : {avg_proba:.3f} (ì•™ìƒë¸”ê³¼ ì¼ì¹˜: {abs(avg_proba - ensemble_proba[1]) < 0.001})")
    
    # ì •ë‹µ ì—¬ë¶€ í™•ì¸
    nb_correct = "âœ…" if actual_label == nb_pred else "âŒ"
    lr_correct = "âœ…" if actual_label == lr_pred else "âŒ"
    ensemble_correct = "âœ…" if actual_label == ensemble_pred else "âŒ"
    
    print(f"ê²°ê³¼: NB {nb_correct} | LR {lr_correct} | ì•™ìƒë¸” {ensemble_correct}")
    
    # ì•™ìƒë¸” íš¨ê³¼ ë¶„ì„
    individual_corrects = sum([actual_label == nb_pred, actual_label == lr_pred])
    if actual_label == ensemble_pred and individual_corrects < 2:
        print("ğŸ¯ ì•™ìƒë¸” íš¨ê³¼: ê°œë³„ ëª¨ë¸ ì‹¤ìˆ˜ë¥¼ ë³´ì™„!")
    elif actual_label != ensemble_pred and individual_corrects == 2:
        print("âš ï¸  ì•™ìƒë¸” ë¶€ì‘ìš©: ê°œë³„ ëª¨ë¸ ì •ë‹µì„ ì˜¤ë‹µìœ¼ë¡œ")

# =============================================================================
# 8. ëª©í‘œ ë‹¬ì„± ì—¬ë¶€ í™•ì¸
# =============================================================================

print(f"\nğŸ¯ ëª©í‘œ ë‹¬ì„± ì—¬ë¶€:")
print("="*60)

target_80 = 0.80
target_81 = 0.81

print(f"ğŸ“Š Soft Voting ì•™ìƒë¸” ìµœì¢… ì„±ëŠ¥: {ensemble_accuracy:.4f}")

if ensemble_accuracy >= target_81:
    print(f"ğŸš€ ìš°ìˆ˜! 81% ëª©í‘œ ë‹¬ì„±: {ensemble_accuracy:.4f} >= {target_81:.4f}")
elif ensemble_accuracy >= target_80:
    print(f"âœ… ì„±ê³µ! 80% ëª©í‘œ ë‹¬ì„±: {ensemble_accuracy:.4f} >= {target_80:.4f}")
    gap_to_81 = target_81 - ensemble_accuracy
    print(f"ğŸ“ˆ 81% ëª©í‘œê¹Œì§€: {gap_to_81:.4f}p ë¶€ì¡±")
else:
    print(f"âš ï¸  80% ëª©í‘œ ë¯¸ë‹¬ì„±: {ensemble_accuracy:.4f} < {target_80:.4f}")
    gap_to_80 = target_80 - ensemble_accuracy
    print(f"ğŸ“ˆ 80% ëª©í‘œê¹Œì§€: {gap_to_80:.4f}p ë¶€ì¡±")

print(f"\nğŸ 1ë‹¨ê³„ Soft Voting ì•™ìƒë¸” ë¶„ì„ ì™„ë£Œ!")
print(f"â° ì™„ë£Œ ì‹œê°„: {datetime.now().strftime('%H:%M:%S')}")

# ê²°ê³¼ ì €ì¥
ensemble_results = {
    'model': soft_voting_clf,
    'cv_scores': cv_scores_ensemble,
    'cv_mean': cv_scores_ensemble.mean(),
    'cv_std': cv_scores_ensemble.std(),
    'val_accuracy': ensemble_accuracy,
    'train_time': train_time_ensemble,
    'cv_time': cv_time_ensemble,
    'predictions': ensemble_predictions,
    'probabilities': ensemble_probabilities,
    'classification_report': ensemble_report
}

print(f"ğŸ’¾ ê²°ê³¼ ì €ì¥ ì™„ë£Œ: ensemble_results ë³€ìˆ˜ì— ì €ì¥ë¨")

# =============================================================================
# 9. ì•™ìƒë¸” ëª¨ë¸ ì €ì¥
# =============================================================================

import joblib

print(f"\nğŸ’¾ ì•™ìƒë¸” ëª¨ë¸ ì €ì¥...")

# ì•™ìƒë¸” ëª¨ë¸ ì €ì¥
joblib.dump(soft_voting_clf, '../../models/trained/ensemble_v1.pkl')

# ë¡œì§€ìŠ¤í‹± íšŒê·€ ëª¨ë¸ë„ ê°œë³„ ì €ì¥ (ë¹„êµìš©)
lr_model = soft_voting_clf.named_estimators_['logistic_regression']
joblib.dump(lr_model, '../../models/trained/logistic_regression_v1.pkl')

print(f"âœ… ì €ì¥ ì™„ë£Œ:")
print(f"   â–ªï¸ ì•™ìƒë¸” ëª¨ë¸: models/trained/ensemble_v1.pkl")
print(f"   â–ªï¸ ë¡œì§€ìŠ¤í‹± íšŒê·€: models/trained/logistic_regression_v1.pkl")


# =============================================================================
# 10. ì•™ìƒë¸” ì˜ˆì¸¡ ê²°ê³¼ ì €ì¥
# =============================================================================

print(f"\nğŸ’¾ ì•™ìƒë¸” ì˜ˆì¸¡ ê²°ê³¼ ì €ì¥...")

import pandas as pd
import os

# results í´ë” ìƒì„± (ì—†ìœ¼ë©´)
os.makedirs('../../results/predictions', exist_ok=True)

# ì•™ìƒë¸” ì˜ˆì¸¡ ê²°ê³¼ DataFrame ìƒì„±
ensemble_predictions_df = pd.DataFrame({
    'actual_label': y_val,
    'predicted_label': ensemble_predictions,
    'predicted_prob_negative': ensemble_probabilities[:, 0],
    'predicted_prob_positive': ensemble_probabilities[:, 1],
    'review_text': X_val,
    'correct_prediction': (y_val == ensemble_predictions).astype(int),
    'confidence': np.max(ensemble_probabilities, axis=1),
    'nb_prediction': nb_results['predictions'],
    'lr_prediction': lr_results['predictions'],
    'nb_prob_positive': nb_results['probabilities'][:, 1],
    'lr_prob_positive': lr_results['probabilities'][:, 1]
})

# í•œê¸€ ì¸ì½”ë”©ìœ¼ë¡œ CSV ì €ì¥
ensemble_predictions_df.to_csv(
    '../../results/predictions/ensemble_predictions.csv', 
    index=False, 
    encoding='utf-8-sig',  # Excelì—ì„œë„ í•œê¸€ì´ ì œëŒ€ë¡œ ë³´ì´ë„ë¡
    quoting=1  # ëª¨ë“  ë¬¸ìì—´ì„ ë”°ì˜´í‘œë¡œ ê°ì‹¸ê¸°
)

print(f"âœ… ì•™ìƒë¸” ì˜ˆì¸¡ ê²°ê³¼ ì €ì¥ ì™„ë£Œ:")
print(f"   â–ªï¸ íŒŒì¼: results/predictions/ensemble_predictions.csv")
print(f"   â–ªï¸ ì´ {len(ensemble_predictions_df):,}ê°œ ì˜ˆì¸¡ ê²°ê³¼")
print(f"   â–ªï¸ ê°œë³„ ëª¨ë¸ ì˜ˆì¸¡ë„ í¬í•¨ (NB, LR)")

# ì•™ìƒë¸” ìƒì„¸ ì„±ëŠ¥ ì§€í‘œ ì €ì¥
ensemble_detailed_metrics = {
    'model_name': 'ensemble_v1',
    'model_type': 'soft_voting_classifier',
    'components': ['naive_bayes', 'logistic_regression'],
    'evaluation_date': datetime.now().isoformat(),
    'validation_performance': {
        'accuracy': float(ensemble_accuracy),
        'precision_positive': float(ensemble_report['ê¸ì •']['precision']),
        'recall_positive': float(ensemble_report['ê¸ì •']['recall']),
        'f1_positive': float(ensemble_report['ê¸ì •']['f1-score']),
        'precision_negative': float(ensemble_report['ë¶€ì •']['precision']),
        'recall_negative': float(ensemble_report['ë¶€ì •']['recall']),
        'f1_negative': float(ensemble_report['ë¶€ì •']['f1-score']),
        'macro_avg_f1': float(ensemble_report['macro avg']['f1-score'])
    },
    'cross_validation': {
        'cv_mean': float(cv_scores_ensemble.mean()),
        'cv_std': float(cv_scores_ensemble.std()),
        'cv_scores': [float(score) for score in cv_scores_ensemble]
    },
    'training_info': {
        'train_time': float(train_time_ensemble),
        'cv_time': float(cv_time_ensemble),
        'voting_type': 'soft',
        'weights': None
    },
    'component_performance': {
        'naive_bayes_accuracy': float(nb_results['val_accuracy']),
        'logistic_regression_accuracy': float(lr_results['val_accuracy']),
        'improvement_over_nb': float(final_nb_improvement),
        'improvement_over_lr': float(final_lr_improvement)
    }
}

with open('../../results/metrics/ensemble_detailed_metrics.json', 'w', encoding='utf-8') as f:
    json.dump(ensemble_detailed_metrics, f, indent=2, ensure_ascii=False)

print(f"   â–ªï¸ ìƒì„¸ ì„±ëŠ¥ ì§€í‘œ: results/metrics/ensemble_detailed_metrics.json")



# =============================================================================
# ì„±ëŠ¥ ê²°ê³¼ JSON ì €ì¥
# =============================================================================

import json

# ì„±ëŠ¥ ì§€í‘œ ì €ì¥
performance_metrics = {
    'naive_bayes': {
        'cv_accuracy_mean': float(nb_results['cv_mean']),
        'cv_accuracy_std': float(nb_results['cv_std']),
        'val_accuracy': float(nb_results['val_accuracy']),
        'train_time': float(nb_results['train_time'])
    },
    'ensemble': {
        'cv_accuracy_mean': float(ensemble_results['cv_mean']),
        'cv_accuracy_std': float(ensemble_results['cv_std']),
        'val_accuracy': float(ensemble_results['val_accuracy']),
        'train_time': float(ensemble_results['train_time'])
    }
}

with open('../../results/metrics/baseline_metrics.json', 'w', encoding='utf-8') as f:
    json.dump(performance_metrics, f, indent=2, ensure_ascii=False)

print(f"âœ… ì„±ëŠ¥ ì§€í‘œ ì €ì¥: results/metrics/baseline_metrics.json")
```
